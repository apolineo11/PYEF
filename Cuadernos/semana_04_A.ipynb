# **Elementos de conteo** <a id="counting_elements"></a>

El objetivo principal del conteo es determinar el número de formas que hay de realizar una operación bien definida.

---

## **Reglas básicas**

### **Regla de la suma**

Si un objeto $A$ puede ser escogido de $m$ posibles formas, y un objeto $B$ puede ser escogido de $n$ posibles formas. Entonces "$A$ *o* $B$" pueden ser escogidos de $m + n$ posibles formas.

### **Regla de la multiplicación**

Si un objeto $A$ puede ser escogido de $m$ posibles formas, y un objeto $B$ puede ser escogido de $n$ posibles formas. Entonces "$A$ *y* $B$" pueden ser escogidos de $m \times n$ posibles formas.

---

**Ejemplo**

Una pareja quiere ir de vacaciones a la playa. Su agencia de viajes les ofrece dos opciones, una vía aerea y otra terrestre. Las aerolineas que viajan de forma directa son 6 y las empresas de transporte terrestre son 3. Sin embargo, dos de las aerolineas no permiten llevar la mascota y una empresa terrestre no cuenta con buena reputación. ¿De cuantas formas puede viajar esta pareja a su destino?

Como 2 de las aerolíneas no permiten llevar mascota, solo quedan 4 disponibles. De igual forma, de las 3 empresas de transporte terrestre, una no es de buena reputación y por ello no está dispuestos a viajar en ella. Entonces la pareja podría viajar de 6 formas posibles.

La pareja ha decidido viajar en una aerolínea que les ofrece tres tipos de guacales para su mascota y tres tipos de sillas para ellos. Con ello, la pareja tiene nueve formas para arreglar sus asientos y el guacal de su mascota.






## **Principio básico del conteo** <a id="basic_principle"></a>

El principo básico del conteo es la misma regla de la multiplicación.

La definición de Ross es la siguiente:

> Supongamos que se van a realizar dos experimentos. Entonces, si el experimento $1$ puede dar uno cualquiera de $m$ resultados posibles y si, para cada resultado del experimento $1$, hay $n$ resultados posibles del experimento $2$, entonces juntos hay $m \times n$ resultados posibles de los dos experimentos.

$$
\begin{array}{cccccc}
(1,1) & (1,2) & \cdots & (1,n) \\
(2,1) & (2,2) & \cdots & (2,n) \\
\vdots & \vdots & \ddots & \vdots \\
(m,1) & (m,2) & \cdots & (m,n) \\
\end{array}
$$

---

## **Principio básico generalizado del conteo** <a id="generalized_basic_principle"></a>

Cuando se tienen más de dos experimentos el principio se puede generalizar así:

> Si $r$ experimentos que se van a realizar son tales que el primero puede dar cualquiera de $n_1$ resultados posibles; y si, para cada uno de estos $n_1$ resultados posibles, hay $n_2$ resultados posibles del segundo experimento; y si, para cada uno de los resultados posibles de los dos primeros experimentos, hay $n_3$ resultados posibles del tercer experimento; y si ..., entonces hay un total de $n_1 \times n2 \times \dots \times n_r$ resultados posibles de los $r$ experimentos.

---

## **Permutaciones** <a id="permutations"></a>

> Una $r$-permutación de $n$ cosas es una selección o disposición *ordenada* de $r$ de ellas.

De acuerdo con esto, la permutación es una arreglo ordenado de los objetos que componen un conjunto. Siguiendo el principio básico de conteo, se tiene que si de un conjunto de tamaño $n$ se van a escoger $r$ elementos de tal forma que es de interés el orden que esta selección tiene. Se puede determinar el número total de las posibles selecciones así:

- Para la primera posición se puede elegir cualquiera de los $n$ elementos, para la segunda posición cualquiera de los $(n-1)$ de los elementos, para la tercera cualquiera de los $(n-2)$ elementos, y así sucesivamente hasta llegar a poder elegir cualquiera de los $(n-r+1)$ elementos restantes.

Cuando $r = n$ se tiene:
$$P(n,n) = n(n-1)(n-2)(n-3) \dots 1$$

Cuando $r \leq n$, se tiene que:

$$
\begin{align*}
P(n,r) &= n(n-1)(n-2)(n-3) \dots (n-r+1)\\
&= \frac{n!}{(n-r)!}
\end{align*}
$$

---

**Ejemplos**

1. En una carrera de fórmula 1 participan 26 corredores. Los cinco primeros ganan puntos según la posición que ocupen (9 puntos al primero, 6 al segundo, etc.) ¿De cuántas maneras pueden repartirse los puntos?

$$
\begin{align*}
P(26, 5) &=  \frac{26!}{(26-5)!}\\
&= 7893600
\end{align*}
$$

2. Un estudiante desea acomodar 4 libros de cálculo, 2 de física, 5 de probabilidad y 3 de álgebra en un estante de tal manera que los libros de la misma materia queden juntos.

$$ 4! \times 4! \times 2! \times 5! \times 3! = 829440$$

3. Se desea calcular el número de maneras  de acomodar a 3 mexicanos, 4 venezolanos, 3 argentinos y 5 colombianos alrededor de una mesa redonda si las personas de la misma nacionalidad insisten en sentarse juntas.

$$ 3! \times 3! \times 4! \times 3! \times 5! = 622080$$

### **Número de permutaciones de $n$ cosas, $p$ de las cuales son de un tipo, $q$ de otro, etc.**

> Supongamos que las $p$ cosas similares se sustituyen por $p$ cosas nuevas, distintas entre sí y de todos los demás tipos de cosas que se permutan, que pueden permutarse de $p!$ maneras; por lo tanto, el número de permutaciones del nuevo conjunto de objetos es $xp!$. Lo mismo ocurre con todos los demás conjuntos de cosas similares y, como finalmente todas las cosas se vuelven diferentes, el número de permutaciones de $n$ cosas diferentes es $n!$.

$$xp!q!\dots = n!$$
$$
\begin{align*}
x &= \frac{n!}{p!q!\dots}
\end{align*}
$$

---

**Ejemplo**

¿Cuántas diferentes disposiciones de las letras pueden formarse con las letras PEPPER?

Primero, observamos que hay $(6!)$ permutaciones de las letras $(P_1E_1P_2P_3E_2R)$ cuando las $(3)$ P y las $(2)$ E son diferenciadas entre sí. Sin embargo, consideremos cualquiera de estas permutaciones, por ejemplo, $(P_1P_2E_1P_3E_2R)$. Si ahora permutamos las $P$ entre ellas y las $E$ entre ellas, el resultado seguirá siendo de la forma PPEPER. Es decir, todas las $(3! \cdot 2!)$ permutaciones son de la forma PPEPER. Por lo tanto, hay $(6! / (3! \cdot 2!) = 60)$ disposiciones posibles de las letras PEPPER.

## **Combinaciones** <a id="combinations"></a>

> Una r-combinación de n cosas es una selección de r de ellas sin tener en cuenta el orden.

> La derivación más sencilla consiste en observar que cada combinación de r cosas distintas puede ordenarse de r! maneras, y así ordenada es una r-permutación.

---

**Ejemplo**

De un grupo de treinta estudiantes queremos escoger dos para participar en una competencia. ¿De cuántas maneras podemos hacerlo?

El primer estudiante del par puede ser cualquiera de los treinta y, una vez escogido éste, el segundo puede ser cualquiera de los veintinueve restantes. Pero de esta manera hemos contado cada pareja dos veces, cuando A es el primero y B el segundo, y cuando B es el primero y A el segundo. Por lo tanto tenemos que dividir este número entre dos.

$$\frac{30 \times 29}{2} = 435$$


---

Del anterior ejemplo se puede derivar la siguiente afirmación:

> Para determinar el número de combinaciones de $C(n,r)$ de orden $r$ de los $n$ objetos, se observa que si se tuviese en cuenta el orden habría $P(n,r)$ de escoger los $r$ objetos, pero como los $r$ objetos pueden permutarse entre si de $r!$ formas entonces:

$$\begin{align*}
r!C(n,r) &= P(n,r) \\
C(n,r) &= \frac{n!}{(n-r)!r!}\\
\binom{n}{r} &= \frac{n!}{(n - r)! \, r!}
\end{align*}$$

Para $0 \leq r\leq n$.

---

**Ejemplos**

1. De un grupo de 10 mujeres y 12 hombres se deben escoger cinco parejas, conformadas por hombre y mujer, para un baile. Se desea determinar el número de selecciones posibles

$$\binom{12}{5} \binom{10}{5} = 199584$$


2. A partir de un grupo de 5 mujeres y 7 hombres, ¿cuántos comités diferentes formados por 2 mujeres y 3 hombres se pueden formar? ¿cuántos si 2 de los hombres están enemistados y se niegan a formar parte juntos del comité?

    - Posibles comités de 2 mujeres y 3 hombres sin restricción.

    $$\binom{5}{2}\binom{7}{3}=350$$

    - Posibles comités de 2 mujeres y 3 hombres sin restricción si dos de los hombres se niegan a estar juntos.

    $$ \left(\binom{2}{1}\binom{5}{2} + \binom{5}{3}\right)\binom{5}{2}= 300$$

## **Coeficiente multinomial** <a id="multinomial_coefficients"></a>


Consideremos el siguiente problema: Un conjunto de $(n)$ elementos distintos se divide en $(r)$ grupos distintos de tamaños respectivos $(n_1, n_2, \ldots, n_r\
)$, donde $(\sum_{i=1}^r n_i = n)$. ¿Cuántas divisiones diferentes son posibles? Para responder a esta pregunta, notamos que:

- Hay $\binom{n}{n_1}$ formas posibles de elegir el primer grupo.
- Para cada elección del primer grupo, hay $\binom{n - n_1}{n_2}$ formas posibles de elegir el segundo grupo.
- Para cada elección de los dos primeros grupos, hay $\binom{n - n_1 - n_2}{n_3}$ formas posibles de elegir el tercer grupo.

Y así sucesivamente. De esto se sigue, a partir de la versión generalizada del principio básico de conteo, se tiene que las posibles divisiones son:

$$
\begin{align}
\binom{n}{n_1} \binom{n - n_1}{n_2} \cdots \binom{n - n_1 - n_2 - \cdots - n_{r-1}}{n_r} &= \\[10pt]
&= \frac{n!}{(n - n_1)! n_1!} \cdot \frac{(n - n_1)!}{(n - n_1 - n_2)! n_2!} \cdots \frac{(n - n_1 - n_2 - \cdots - n_{r-1})!}{0! n_r!} \\[10pt]
&= \frac{n!}{n_1! n_2! \cdots n_r!}
\end{align}
$$

---

**Ejemplo**

1. Los treinta alumnos de tercer grado de una escuela deben ser repartidos en 5 grupos con 12, 5, 3, 6 y 4 alumnos respectivamente. ¿Cuántas reparticiones son posibles?

$$\binom{30}{12,5,3,6,4}=\frac{30!}{12! \times 5! \times 3! \times 6! \times 4!}=4.4509 \times 10^{16}$$

## **Número de soluciones enteras de una ecuación** <a id="multinomial_coefficients"></a>

El siguiente texto es tomado del libro de Liliana Blanco, 2004.

> Supóngase que se tiene $n$ bolas indistinguibles entre ellas.
¿De cuántas formas se pueden distribuir las $n$ bolas en $r$ urnas?. \\
El resultado de este experimento puede ser descrito por un vector
$(x_1, \cdots, x_r)$, donde $x_i$ denota el número de bolas colocadas en la urna $i$. Por lo tanto, el problema se reduce a encontrar el número de vectores $(x_1, \cdots, x_r)$, con componentes en los enteros no negativos, tales que $x_1 + \cdots + x_r = n$. \\
Para resolver este problema, supóngase que se colocan los objetos en
una línea horizontal y que se dividen en $r$ grupos de la siguiente forma:
hay $(n-1)$ espacios entre los objetos, se escogen $(r-1)$ de esos espacios y
allí se trazan líneas divisorias, al hacerlo quedan $r$ grupos ninguno de ellos
vacío. Esto es, hay $\binom{n-1}{r-1}$ vectores $(x_1, \cdots, x_r)$, con componentes positivas que satisfacen $x_1 + \cdots + x_r = n$. \\
Para hallar el número de soluciones no negativas, se observa que el
número de soluciones no negativas de $x_1 + \cdots + x_r = n$, es igual al número de soluciones positivas de $y_1 + \cdots + y_r = n + r$, con $y_i = x_i + 1$ para $i = 1, 2, \cdots, r$. \\
Por lo tanto, el número total de formas en que se pueden distribuir las
$n$ bolas indistinguibles en $r$ urnas es:

$$\binom{n + r - 1}{r - 1}$$

Con lo anterior queda mostrado como se deriva el número de soluciones no negativas de una ecuación.

---

**Ejemplo**

1. Se reparten 12 regalos entre 7 niños. ¿Cuántas reparticiones son posibles ?, ¿Cuántas si cada niño debe recibir por lo menos un regalo?

    - $$\binom{12 + 7 - 1}{7 - 1} = 18564$$
    - $$\binom{12 + 1}{7 - 1} = 462$$

## **Elementos de conjuntos** <a id="set_theory_elements"></a>


> Un conjunto es una agrupación de objetos bien definidos. Los objetos que integran un conjunto son llamados elementos o miembros del conjunto. (Blanco, 2004)

Los elementos acostumbran a nombrar con letras latinas mayúsculas $A, B, C, \dots$, mientras que los elementos de cada conjunto con letras latinas minísculas $a, b, c, \dots$

- Los conjuntos pueden ser definidos de dos formas principalmente:

    - Extensión: $$A = {1,2,6}$$
    - Comprensión: $$B=\{x : x \bmod 2 = 0\}$$

- Se nota que un elemento $a$ hace parte del conjunto A así: $a \in A$.
- Un conjunto que su cardinal se puede expresar como un entero no negativo se llama conjunto finito
- El conjunto que no tiene elementos se llama conjunto vacio y se nota $\Phi$.
- Dos conjuntos son iguales, si y sólo si, ellos tienen exactamente los mismos elementos.
- Si todos los elementos de un conjunto $A$ son elementos de un conjunto $B$, se dice que $A$ está contenido en $B$. Y se nota $A \subseteq B$
- Si $A$ está contenido en $B$ pero por lo menos hay un elemento de $B$ que no está en $A$, se dice que $A$ es subconjunto de $B$ y se nota $A \subset B$.
- Todo conjunto es subconjunto de si mismo
- El conjunto vacio $\Phi$ es subconjunto de todo conjunto. Como $\Phi$ no tiene elementos no puede violar la definición de subconjunto.
- En la teoría todos los conjuntos son subconjuntos de un mismo subconjunto $\mathcal{U}$.


### **Unión**

La uníon de conjuntos $A \cup B$ es el conjunto en donde todos sus elementos pertenecen al conjunto A o al conjunto $B$ o a ambos.

$$A \cup B := \{x:x \in A \lor x \in B\}$$

### **Intersección**

La intersecciónm de dos conjuntos $A \cap B$, es el conjunto de los elementos que pertenecen a A y B.

$$A \cap B := \{x:x \in A \land x \in B\}$$


### **Diferencia**

La diferencia de dos conjuntos $A-B$ es el conjunto de elementos que pertenencen a $A$ y que no pertenencen a $B$.

$$A - B := \{x:x \in A \land x \notin B\}$$

### **Diferencia simétrica**

La diferencia simétrica de dos conjuntos $A \triangle B$, es el conjunto de los elementos que pertenecen a $A$ o $B$ pero que no pertenecen a su intersección.

$$A \triangle B := (A \cup B) - (A \cap B)$$


### **Complemento**

El complemento de un conjunto $A$ es el conjunto de elementos que no pertenecen a $A$ y que pertenencen al conjunto universal $\mathcal{U}$.


$$A^c := \{x: x \in \mathcal{U} \land x \notin  A\}$$


### **Propiedades y leyes**

1. **Leyes conmutativas**

    a. $A \cup B = B \cup A$

    b. $A \cap B = B \cap A$  

2. **Leyes asociativas**

    a. $A \cup (B \cup C) = (A \cup B) \cup C$

    b. $A \cap (B \cap C) = (A \cap B) \cap C$

3. **Leyes distributivas**

    a. $A \cap (B \cup C) = (A \cap B) \cup (A \cap C)$

    b. $A \cup (B \cap C) = (A \cup B) \cap (A \cup C)$

4. **Leyes complementarias**  

    a. $A \cup A^c = \mathcal{U}$

    b. $A \cap A^c = \Phi$

    c. $A \cup \mathcal{U} = \mathcal{U}$

    c. $A \cap \mathcal{U} = A$

    d. $A \cup \Phi = A$

    f. $A \cap \Phi = \Phi$

5. **Leyes de la diferencia**

    a. $A - B = A \cap B^c$

    b. $A - B = A - (A \cap B) = (A \cup B) - B$

    c. $A - (B - C) = (A - B) \cup (A \cap C)$

    d. $(A \cup B) - C = (A - C) \cup (B - C)$

    e. $(A \cap B) - C = (A - C) \cap (B - C)$

    f. $(A \cap B) \cup (A - B) = A$

    g. $(A \cap B) \cap (A - B) = \Phi$


6. **Leyes de De-Morgan**

    a. $(A \cap B)^c = A^c \cup B^c \)$
    
    b. $(A \cup B)^c = A^c \cap B^c \)$

7. **Ley involutiva**

   $(A^c)^c = A$

8. **Leyes de idempotencia**

    a. $A \cup A = A$

    b. $A \cap A = A$
