# Proceso de Poisson

Un **Proceso de Poisson** es un modelo estocástico que describe la ocurrencia de eventos aleatorios a lo largo del tiempo. Se utiliza en situaciones donde los eventos ocurren de manera independiente y a una tasa constante.

---

## Definición
Un proceso de Poisson {N(t), t ≥ 0} es un **proceso de conteo** que satisface las siguientes condiciones:

1. **N(0) = 0**  
   - El proceso comienza en t = 0 sin eventos.

2. **Incrementos Independientes**  
   - El número de eventos en intervalos disjuntos es independiente.  
   - Ejemplo: El número de llamadas entre 12:00-12:30 es independiente del número de llamadas entre 13:00-13:30.

3. **Incrementos Estacionarios**  
   - La distribución de N(t + s) - N(t) solo depende de s, no del tiempo inicial t.  
   - Es decir, el número de eventos en un intervalo de 10 minutos es el mismo sin importar la hora en la que empiece.

4. **Distribución de los Incrementos Pequeños**  
   - Para un intervalo pequeño h:
     - La probabilidad de un evento en [t, t+h] es aproximadamente P(N(t+h) - N(t) = 1) ≈ λh.
     - La probabilidad de dos o más eventos es despreciable: P(N(t+h) - N(t) ≥ 2) ≈ 0.

---

## Distribución del Número de Eventos
El número de eventos en un intervalo de longitud t sigue una **distribución de Poisson** con parámetro λt:

P(N(t) = k) = [(λt)^k * e^(-λt)] / k!,  para k = 0,1,2, ...

Donde:
- λ es la **tasa de ocurrencia** (número esperado de eventos por unidad de tiempo).
- t es la duración del intervalo de observación.

---

## Propiedades Claves
- **Tiempo entre eventos**:  
   - Sigue una distribución **Exponencial** con parámetro λ.  
   - T ~ Exp(λ), donde su función de densidad es f_T(t) = λ e^(-λt).  
   - Esto implica que el proceso es **sin memoria** (la probabilidad de un evento futuro no depende del pasado).

- **Superposición y División**:  
   - Si hay varios procesos de Poisson con tasas λ1, λ2, ...,  
     la suma sigue un **Poisson con tasa λ = λ1 + λ2 + ...**.  
   - Si cada evento de un proceso de Poisson con tasa λ se clasifica en k tipos con probabilidades p1, p2, ..., pk,  
     los eventos en cada categoría siguen un proceso de Poisson con tasas λpi.

---

## Ejemplo
Si en un **centro de llamadas** llegan en promedio λ = 5 llamadas por minuto:

- ¿Cuál es la probabilidad de recibir exactamente 3 llamadas en un minuto?

$P(N(1) = 3) = [(5)^3 * e^(-5)] / 3! ≈ 0.1404.$

- ¿Cuál es el tiempo esperado entre llamadas?

$E[T] = 1/λ = 1/5 = 0.2$ minutos = 12 segundos.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
lambda_poisson = 5  # tasa de eventos por unidad de tiempo
T = 10  # tiempo total de simulación
n_eventos = np.random.poisson(lambda_poisson * T)

# Generar tiempos entre eventos (distribución exponencial)
event_times = np.cumsum(np.random.exponential(1/lambda_poisson, n_eventos))

# Graficar
plt.figure(figsize=(10,4))
plt.step(event_times, np.arange(1, n_eventos+1), where="post", label="Proceso de Poisson")
plt.xlabel("Tiempo")
plt.ylabel("Número de eventos")
plt.title("Simulación de un Proceso de Poisson")
plt.legend()
plt.grid()
plt.show()
```
