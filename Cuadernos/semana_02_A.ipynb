## **Tablas de frecuencias**

### Tabla de frecuencias absolutas

La tabla de frecuencias absolutas registra los **conteos** de los niveles de **UNA** variable categórica. En la columna de la izquierda ponemos los niveles de la variable categórica (valores posibles $[1,k]$), y en la columna de la derecha los conteos (frecuencia absoluta) de la variable en dicho nivel. La suma de todas la frecuencias absolutas debe ser igual al tamaño de la muestra (número de registros en los datos).


---



|Niveles de la variable | Frecuencia absoluta |
| :-------------------: | :-----------------: |
| $l_{1}$               | $f_{1}$             |
| $l_{2}$               | $f_{2}$             |
| $\vdots$              | $\vdots$            |
| $l_{k}$               | $f_{k}$             |
| **TOTAL**             | $\sum_{1}^{k} f_{l}= n$|   



---


### Tabla de frecuencias relativas

La tabla de frecuencias relativas es muy similar a la tabla de frecuencias absolutas. En esta tabla registramos las **proporciones** de cada uno de los niveles de la variable categórica con respecto al número total de registros ($n$).



---

|Niveles de la variable | Frecuencia relativa |
| :-------------------: | :-----------------: |
| $l_{1}$               | $\frac{f_{1}}{n}$             |
| $l_{2}$               | $\frac{f_{2}}{n}$             |
| $\vdots$              | $\vdots$            |
| $l_{k}$               | $\frac{f_{k}}{n}$             |
| **TOTAL**             | $\frac{1}{n}\sum_{1}^{k} f_{l}= 1$|  

## Tablas cruzadas o de contigencia <a id="tabla_contingencia"></a>

Las tablas de contingencia de permiten registrar las **cantidades** y/o las **proporciones** de **DOS** variables categóricas.

- Sean $C$ y $D$ dos variables categóricas cada una de las cuales tiene $k$ y $l$ niveles respectivamente.

- Se nota $\#(C_{i} \cap C_{j})$ al total de registros (frecuencia) que en la categoría $C$ tienen el nivel $i$ y en la categoría $D$ tienen el nivel $j$ de forma simultanea.

---

|           |    | Tabla de Contingencia |         |        |**Total**|
|:---------:|:---------------------: |:----------------------:|:--------:|:------:|:------:|
|           | $D_{1}$                | $D_{2}$                | $\dots$  |$D_{j}$|-------------|
| $C_{1}$   | $\#(C_{1} \cap D_{1})$ | $\#(C_{1} \cap D_{2})$ | $\dots$  | $\#(C_{1} \cap D_{j})$ |$\#(C_{1})$|
| $C_{2}$   | $\#(C_{2} \cap D_{1})$ | $\#(C_{2} \cap D_{2})$ | $\dots$  | $\#(C_{2} \cap D_{j})$ |$\#(C_{2})$|
| $\vdots$  | $\vdots$              | $\vdots$              | $\ddots$   | $\vdots$ |$\vdots$|
| $C_{k}$   | $\#(C_{k} \cap D_{1})$ | $\#(C_{k} \cap D_{2})$ | $\dots$   | $\#(C_{k} \cap D_{j})$|$\#(C_{k})$|
| **Total** |     $\#(D_{1})$        |     $\#(D_{2})$        | $\dots$  | $\#(D_{j})$ |$n$|

---

## Tabla de proporciones condicional <a id="tabla_condicional"></a>

Muchas veces requerimos contestar preguntas en el que buscamos  la diferencia entre los niveles de una categoría en términos de otra. Para ello usamos como base la tabla de contigencias y dividimos cada celda por el total de su respectiva columna.



---

|           |    | Tabla de proporciones condicional |         |        |**Total**|
|:---------:|:---------------------: |:----------------------:|:--------:|:------:|:------:|
|           | $D_{1}$                | $D_{2}$                | $\dots$  |$D_{j}$|-------------|
| $C_{1}$   | $\frac{\#(C_{1} \cap D_{1})}{\#(C_{1})}$ | $\frac{\#(C_{1} \cap D_{2})}{\#(C_{1})}$ | $\dots$  | $\frac{\#(C_{1} \cap D_{j})}{\#(C_{1})}$ |$1$|
| $C_{2}$   | $\frac{\#(C_{2} \cap D_{1})}{\#(C_{2})}$ | $\frac{\#(C_{2} \cap D_{2})}{\#(C_{2})}$ | $\dots$  | $\frac{\#(C_{2} \cap D_{j})}{\#(C_{2})}$ |$1$|
| $\vdots$  | $\vdots$              | $\vdots$              | $\ddots$   | $\vdots$ |$\vdots$|
| $C_{k}$   | $\frac{\#(C_{k} \cap D_{1})}{\#(C_{k})}$ | $\frac{\#(C_{k} \cap D_{2})}{\#(C_{k})}$ | $\dots$   | $\frac{\#(C_{k} \cap D_{j})}{\#(C_{k})}$|$1$|

---

Si $\#(C_{k}) = 0$ entonces el valor de la celda se deja en cero $(0)$.

## Mercado inmobiliario de Melbourne

### Variables

- **Suburb**: Suburb

- **Address**: Address

- **Rooms**: Number of rooms

- **Price**: Price in Australian dollars

- **Method**:

    - SP: property sold prior

    - br: bedroom(s)

    - u: unit, duplex
    - t: townhouse
    - dev site: development site
    - res: other residential

- **SellerG**: Real Estate Agent

- **Date**: Date sold

- **Distance**: Distance from CBD in Kilometres

- **Regionname**: General Region (West, North West, North, North east …etc)

- **Propertycount**: Number of properties that exist in the suburb.

- **Bedroom2**: Scraped # of Bedrooms (from different source)

- **Bathroom**: Number of Bathrooms

- **Car**: Number of carspots

- **Landsize**: Land Size in Metres

- **BuildingArea**: Building Size in Metres

- **YearBuilt**: Year the house was built

- **CouncilArea**: Governing council for the area

- **Lattitude**: Self explanitory

- **Longtitude**: Self explanitory



```python
# https://www.kaggle.com/code/janiobachmann/melbourne-comprehensive-housing-market-analysis
# https://www.kaggle.com/datasets/anthonypino/melbourne-housing-market

import pandas as pd
import numpy as np

# data URL
url = "https://raw.githubusercontent.com//apolineo11/PYEF/main/Cuadernos/Data/Melbourne_housing_FULL.csv"


# Load the data into a DataFrame
data_housing = pd.read_csv(url, parse_dates=['Date'], date_format='%d/%m/%Y')#.dropna().reset_index(drop=True)

# Display the first few rows
data_housing.iloc[list(np.random.randint(0,data_housing.shape[0],10))]
```

```python
###########################
## Tablas de frecuencias ##
###########################

# convertir las columnas en columnas categóricas
method_categories = ['S', 'SP', 'PI', 'PN', 'SN', 'NB', 'VB', 'W', 'SA', 'SS', 'N/A']
type_categories = ['br', 'h', 'u', 't', 'dev site', 'res']


data_housing['Method'] = pd.Categorical(data_housing['Method'], categories=method_categories, ordered=False)
data_housing['Type'] = pd.Categorical(data_housing['Type'], categories=type_categories, ordered=False)
```

```python
# Combined absolute frequency and relative frequency table for Method column
method_counts = data_housing['Method'].value_counts()
method_proportions = data_housing['Method'].value_counts(normalize=True)

method_frequency_table = pd.DataFrame({
    'Count': method_counts,
    'Proportion': method_proportions
}).reset_index()
print('------------------------')
print(method_frequency_table)
print('------------------------')
print(method_frequency_table[['Count', 'Proportion']].sum())
```

```python
# Combined absolute frequency and relative frequency table for Type column
type_counts = data_housing['Type'].value_counts()
type_proportions = data_housing['Type'].value_counts(normalize=True)

type_frequency_table = pd.DataFrame({
    'Count': type_counts,
    'Proportion': type_proportions
}).reset_index()
print('------------------------')
print(type_frequency_table)
print('------------------------')
print(type_frequency_table[['Count', 'Proportion']].sum())
```

```python
#######################################
## Tablas cruzadas o de contingencia ##
#######################################

df_cross = data_housing.groupby(['Method', 'Type']).size().reset_index(name='Counts')
df_cross = pd.pivot_table(df_cross, index='Method', columns='Type', values='Counts', aggfunc='sum', margins=True,  margins_name='Total', observed=True)
df_cross.reset_index(inplace=True)
df_cross.columns.name = None
df_cross.index.name = None
df_cross
```

```python
condional_propotion_columns = {}
for column in type_categories:
    condional_propotion_columns[column] = list(df_cross[column]/df_cross['Total'])[:-1]

condional_propotion_columns['Method'] = list(df_cross['Method'])[:-1]
df_conditional_proportion = pd.DataFrame(condional_propotion_columns).fillna(0)
df_conditional_proportion['Total'] = df_conditional_proportion[['br', 'h', 'u', 't', 'dev site', 'res']].sum(axis=1)
cols = ['Method'] + [col for col in df_conditional_proportion.columns if col != 'Method']
df_conditional_proportion = df_conditional_proportion[cols]
df_conditional_proportion
```

```python
# Gráfico de barras
import plotly.express as px

# Barras para variable Method
# Frecuencia absoluta
fig = px.bar(method_frequency_table, x='Method', y='Count', title='Frecuencia Absoluta para la variable Method')
fig.update_yaxes(title="Frecuncia Absoluta")
fig.show()
```

```python
# Frecuencia relativa
fig = px.bar(method_frequency_table, x='Method', y='Proportion', title='Frecuencia relativa para la variable Method')
fig.update_yaxes(title="Frecuncia Relativa")
fig.show()
```

```python
# Barras para variable Method
# Frecuencia absoluta
fig = px.bar(type_frequency_table, x='Type', y='Count', title='Frecuencia Absoluta para la variable Type')
fig.update_yaxes(title="Frecuncia Absoluta")
fig.show()
```

```python
# Frecuencia relativa
fig = px.bar(type_frequency_table, x='Type', y='Proportion', title='Frecuencia realtiva para la variable Type')
fig.update_yaxes(title="Frecuncia relativa")
fig.show()
```

```python
# Barras apiladas
data_grouped_method = data_housing.groupby(['Method', 'Type']).size().reset_index(name='Counts')
fig = px.bar(data_grouped_method, x="Method", y="Counts", color="Type", text_auto=True, title='Distribución de la variable Type para las categorías de la Variable Método')
fig.show()
```

```python
fig = px.bar(data_grouped_method, x="Type", y="Counts", color="Method", text_auto=True, title='Distribución de la variable Method para las categorías de la Variable Type')
fig.show()
```

```python
# Barras proporciones condicionadas
df_conditional_proportion_long = pd.melt(df_conditional_proportion[['Method','br','h','u','t','dev site','res']], id_vars='Method', var_name='Type', value_name='conditional_proportion')
fig = px.bar(df_conditional_proportion_long, x="Method", y="conditional_proportion", color="Type", text_auto=True, title='Proporciones condicionales de la variable Type para las categorías de la Variable Método')
fig.show()
```
