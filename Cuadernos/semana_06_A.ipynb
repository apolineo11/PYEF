## **Conceptos** <a id="concepts"></a>

En ocasiones se obtiene informacón parcial de un experimento aleatorio *antes* de que se conozca su resultado final. Este *conocimiento* previo provoca un cambio en la estructura probabilistica de los resultados.

**Ejemplo**

- En un juego de poker (Texas Hold Them), uno de los jugadores accidentalmente ha dejado ver unas de sus cartas, dicha carta es un Rey de picas. Dado que en las cartas comunes (Flop) hay 3 picas, la propabilidad que el jugador del Rey de picas obtenga color alto ha aumentado. En razón ha esto, los jugadores que alcanzaron a ver su carta han decidido abandonar.

- Las probabilidades que una persona tenga complicaciones si ha contraido una gripe son bajas. Pero si dicha persona tiene complicaciones con su sistema inmune, las probabilidades de complicaciones cambian.

## Frecuencia relativa y probabilidad condicional

En términos generales, la aproximación al concepto de probabilidad condicional se hace desde la *frecuencia relativa*.

Sea:

- $A$: El evento que ya ha sido observado
- $B$: El evento no observado y del que tenemos interés en su estructura probabilistica.


Si $n(A) > 0$ se tiene que:

$$
\begin{align*}
f_{r}(B|A) &= \frac{n(B \cap A)}{n(A)}\\
&=\frac{\frac{n(B \cap A)}{n}}{\frac{n(A)}{n}}\\
&=\frac{f_{r}(B \cap A)}{f_{r}(A)}
\end{align*}$$

<br>

La expresión $n(B \cap A)$, se debe leer como el número de veces en que sucedieron los dos eventos. Y al igual que antes, si el experimento se realiza un número *suficiente* de veces el núimero $f_{r}(B|A)$ tiende a estabilizarce en $P(B|A)$ y $f_{r}(A)$ en $P(A)$.

## Definición de probabilidad condicional

- Sea $(\Omega, \mathcal{F}, P)$ un espacio de probabilidad.
- $A, B \in \mathcal{F}$.
- $P(A) > 0$

<br>

Entonces:

$$P(B|A) = \frac{P(B \cap A)}{P(A)}$$


<br>

---

<br>

**Ejemplo**

Se lanzan 2 dados corrientes una vez.
Nos interesa saber la probabilidad de que al menos uno de los resultados sea 6 dado que los resultados son diferente.

- $A:=\lbrace (a, 6): a \in \lbrace 1,2,3,4,5,6 \rbrace \rbrace \cup \lbrace (6, a): a \in \lbrace 1,2,3,4,5,6 \rbrace \rbrace$

- $B:=\lbrace (a,b): a \neq b \in  \lbrace 1,2,3,4,5,6 \rbrace \rbrace$


Si suponemos un espacio laplaciano se tiene que:

- $P(B)=\frac{30}{36}$
- $P(A \cap B) = \frac{10}{36}$
- $P(A|B) = \frac{P(A \cap B)}{P(B)} = \frac{1}{3}$

## **Medida de probabiliad condicional** <a id="conditional_probability_measure"></a>

Sea $(\Omega, \mathcal{F}, P)$ un espacion de probabilidad y $A \in \mathcal{F}$ con $P(A) > 0$. Entonces:

1. $P(\cdot \mid A)$ es una medida de probabilidad sobre $\Omega$, que está concentrada en $A$, esto es $P(A \mid A) = 1$.

2. Si $A \cap B = \varnothing$ entonces $P(B \mid A) = 0$.

3. $P(B \cap C \mid A) = P(B \mid A \cap C) P(C \mid A)$, si $P(A \cap C) > 0$.

4. Si $A_1, A_2, \dots, A_n \in \mathcal{S}$ con $P(A_1 \cap A_2 \cap \cdots \cap A_{n-1}) > 0$ entonces  
   $$
   P(A_1 \cap A_2 \cap \cdots \cap A_n) = P(A_1) P(A_2 \mid A_1) P(A_3 \mid A_1 \cap A_2) \cdots P(A_n \mid A_1 \cap A_2 \cap \cdots \cap A_{n-1}).
   $$

<br>

*Revisar la demostración en Blanco Castañeda, L. (2004). Probabilidad. Universidad Nacional de Colombia - Facultad de Ciencias*

<br>

---

**Ejemplo**

Tres adolescentes desean entrar a una película para mayores de 18 años. El encargado de la taquilla, para verificar la edad, les pide su documento de identidad. Los jóvenes lo entregan y el dependiente, luego de revisarlos y negarles la entrada, los devuelve al azar. Calcular la probabilidad de que ninguno de los jóvenes reciba su propio documento.

*Solución:* Sean  

$$
A := \text{"ningún joven recibe su propio documento"}
$$
$$
B_i := \text{"el i-ésimo joven recibe su propio documento"}.
$$

Es claro que la probabilidad buscada es:

$$
\begin{align*}
P(A) &= P(B_1^c \cap B_2^c \cap B_3^c)\\
&= 1 - P(B_1 \cup B_2 \cup B_3)\\
&= 1 - P(B_1) - P(B_2) - P(B_3) + P(B_1 \cap B_2) + P(B_2 \cap B_3) + P(B_1 \cap B_3) - P(B_1 \cap B_2 \cap B_3)\\
\end{align*}
$$

Como hay tres casos posibles y uno favorable, se tiene que:  
$$
P(B_i) = \frac{1}{3} \quad \text{para } i = 1, 2, 3.
$$


Por otra parte,

$$
P(B_i \cap B_j) = P(B_i)P(B_j \mid B_i) = \frac{1}{3} \times \frac{1}{2} = \frac{1}{6},
$$

ya que al haber entregado al $i$-ésimo joven correctamente su documento de identidad, para el $j$-ésimo queda sólo una opción favorable entre dos posibles. Análogamente se verifica que

$$
P(B_1 \cap B_2 \cap B_3) = P(B_1)P(B_2 \mid B_1)P(B_3 \mid B_1 \cap B_2)
= \frac{1}{3} \times \frac{1}{2} \times 1 = \frac{1}{6}.
$$

Por lo tanto,

$$
P(A) \approx 0.333.
$$

## **Probabilidad Total** <a id="total_probability"></a>

Sea $A_1, A_2, \dots$ una partición finita o numerable de $\Omega$, es decir, $A_i \cap A_j = \varnothing$ para todo $i \neq j$ y  

$$
\bigcup_{i=1}^\infty A_i = \Omega,
$$

tal que $P(A_i) > 0$ para todo $i$. Entonces, para cualquier $B \in \mathcal{S}$ se satisface:  

$$
P(B) = \sum_n P(B \mid A_n)P(A_n).
$$

**Demostración.**  

$$
P(B) = P(B \cap \Omega)
$$

$$
= P\left(\bigcup_n (B \cap A_n)\right)
$$

$$
= \sum_n P(B \cap A_n)
$$

$$
= \sum_n P(B \mid A_n)P(A_n).
$$


## **Regla de Bayes** <a id="bayes_rule"></a>

Sea $A_1, A_2, \dots$ una partición finita o numerable de $\Omega$ con $P(A_i) > 0$ para todo $i$, entonces se satisface para todo $B \in \mathcal{S}$ con $P(B) > 0$ lo siguiente:

$$
P(A_i \mid B) = \frac{P(A_i)P(B \mid A_i)}{\sum_j P(B \mid A_j)P(A_j)}, \quad \text{para todo } i.
$$

**Demostración.**

$$
\begin{align*}
P(A_i \mid B) =& \frac{P(A_i \cap B)}{P(B)}\\
=& \frac{P(A_i)P(B \mid A_i)}{P(B)}\\
=& \frac{P(A_i)P(B \mid A_i)}{\sum_j P(B \mid A_j)P(A_j)}\\
\end{align*}
$$


> *Para dar una interpretación a la regla de Bayes, supóngase que los eventos $A_1, A_2, \dots$ son todas las posibles causas, mutuamente excluyentes, de un evento $B$. Bajo el supuesto que el evento $B$ ha sido observado, la fórmula de Bayes permite conocer cuál de estas causas es la más probable de haber producido el evento $B$*. (Blanco Castañeda, L. (2004))


<br>

**Ejemplo**  
La cuarta parte de una población está vacunada contra una enfermedad contagiosa. En el transcurso de una epidemia debida a tal enfermedad, se observa que de cada cinco enfermos sólo uno está vacunado. Se sabe además que de cada doce vacunados sólo uno está enfermo. Se quiere calcular la probabilidad de que un no vacunado esté enfermo.

Sean:

$$
V := \text{"La persona está vacunada"}, \quad
E := \text{"La persona está enferma"}.
$$

De acuerdo a la información suministrada se tiene que:

$$
P(V \mid E) = \frac{1}{5}, \quad
P(E \mid V) = \frac{1}{12}, \quad
P(V) = \frac{1}{4}.
$$

Por lo tanto,

$$
\begin{align*}
x &:= P(E \mid V^c) \\
&= \frac{P(V^c \mid E) P(E)}{P(V^c)} \\
&= \frac{\frac{4}{5} \left( \frac{1}{12} \cdot \frac{1}{4} + \frac{3}{4}x \right)}{\frac{3}{4}},
\end{align*}
$$

de donde se obtiene

$$
x = \frac{1}{9}
$$



## **Eventos independientes** <a id="independent_events"></a>

En muchas ocasiones la ocurrencia de un evento $(B)$ no afecta la probabilidad de ocurrencia de otro evento $(A)$.

$$P(A|B) = P(A)\text{, si } P(B)>0.$$


Se dice que el evento $A$ es independiente del evento $B$.

**Definición**

Dos eventos $A$ y $B$ son independientes, si y solo si,

$$P(A \cap B) = P(A)P(B)$$

En caso contrario lo eventos son dependientes.

<br>

*Consideración: Los eventos mutuamente excluyentes NO son necesariamente independientes.*

<br>

---

<br>

**Ejemplos**

1. Supóngase que se lanza un dado corriente dos veces consecutivas. Sean:

$$
A := \text{"la suma de los resultados obtenidos es un número par"}, \quad
B := \text{"el resultado del segundo lanzamiento es par"}.
$$

En este caso:

$$
\begin{align*}
P(A) &= P(B) = \frac{1}{2}, \\
P(A \cap B) &= \frac{1}{4}.
\end{align*}
$$

Por lo tanto, los eventos son independientes

<br>

---

<br>


2. Se carga un dado de manera que la probabilidad de obtener un número par es igual a $\frac{2}{5}$. Sean $A$ y $B$ como en el ejemplo anterior. En este caso se tiene que:

$$
\begin{align*}
P(A) &= \frac{13}{25}, \\
P(B) &= \frac{2}{5}, \\
P(A \cap B) &= \frac{4}{25}.
\end{align*}
$$

Por lo tanto, $A$ y $B$ no son independientes.
