# Desigualdad de Chebyshev Unilateral <a id="chebyshev_onesided"></a>

Sean $\bar{x}$ y $s$ la media muestral y la desviación estándar muestral de los datos $X_{n}=\{x_{1}, x_{2}, \dots, x_{n}\}$. Además, suponga que $s > 0$.

Entonces, sea $N(k) = \{x_{i}:x_{i}-\bar{x} \geq ks\}$, si $|N(k)|$ es el ordinal del conjunto $N(k)$, se tiene que para cualquier $k>0$:

$$\frac{|N(k)|}{n} \leq \frac{1}{1+k^{2}}$$

---

## Prueba de la desigualdad de Chebyshev Unilateral

### Parte 1

Sea $(y_{i} = x_{i} - \bar{x}), ( i = 1, \ldots, n )$. Para cualquier $( b > 0 )$, tenemos que:

$$
\begin{align*}
\sum_{i=1}^{n} (y_i + b)^2 &\geq \sum_{i: y_i \geq ks} (y_i + b)^2 \\
&\geq \sum_{i: y_i \geq ks} (ks + b)^2 \\
&= |N(k)| (ks + b)^2
\end{align*}
$$

---

### Parte 2

$$
\begin{align*}
\sum_{i=1}^{n} (y_i + b)^2 &= \sum_{i=1}^{n} (y_i^2 + 2b y_i + b^2) \\
&= \sum_{i=1}^{n} y_i^2 + 2b \sum_{i=1}^{n} y_i + n b^2 \\
&= \sum_{i=1}^{n} y_i^2 + n b^2 \\
&= (n - 1) s^2 + n b^2
\end{align*}
$$

---

### Parte 3

$$\begin{align*}
|N(k)| &\leq \frac{(n - 1)s^2 + n b^2}{(ks + b)^2}
< \frac{n s^2 + n b^2}{(ks + b)^2}
\end{align*}
$$

---


### Parte 4

$$
\begin{align*}
\frac{|N(k)|}{n} &\leq \frac{s^2 + b^2}{(ks + b)^2}
\end{align*}
$$

---

### Parte 5

Primero reemplazamos $b = \frac{s}{k}$, y luego en el lado derecho de la desigualdad multiplicamos un uno de la forma $\frac{k^{2}}{s^{2}}$.

$$\begin{align}
\frac{|N(k)|}{n} &\leq \frac{s^2 + s^2 / k^2}{(ks + s / k)^2} \\
&\leq \frac{k^2 + 1}{(k^2 + 1)^2} = \frac{1}{k^2 + 1}
\end{align}
$$

# Medidas de las relaciones para dos variables cuantitativas <a id="covariance_correlation"></a>

> Una cuestión de interés relativa a los conjuntos de datos emparejados es si los valores grandes de $x$ tienden a emparejarse con valores grandes de $y$, y los valores pequeños de $x$ con valores pequeños de $y$; si no es así, entonces podríamos preguntarnos si los valores grandes de una de las variables tienden a emparejarse con valores pequeños de la otra. (Ross, 2022)

---

## Covarianza

La covarianza $(Cov)$ es una medida de la relación lineal entre dos variables. Un valor positivo indica una relación lineal directa o creciente y un valor negativo indica una relación lineal decreciente

### Covarianza poblacional

$$
\text{Cov}(x, y) = \sigma_{xy} = \frac{\sum_{i=1}^{N} (x_i - \mu_x)(y_i - \mu_y)}{N}
$$

### Covarianza muestral


$$
\text{Cov}(x, y) = s_{xy} = \frac{\sum_{i=1}^{N} (x_i - \bar{x})(y_i - \bar{y})}{n-1}
$$

---

## Correlación de Pearson


El coeficiente de correlación de Pearson se define como:

$$
r(X,Y) = \frac{\sum_{i=1}^{n} (x_{i} - \bar{x})(y_{i} - \bar{y})}{\sqrt{\sum_{i=1}^{n} (x_{i} - \bar{x})^2} \sqrt{\sum_{i=1}^{n} (y_{i} - \bar{y})^2}}
$$

donde:
- $(y_{i})$ y $(y_{i})$ son los valores de las variables $(X)$ y $(Y)$.
- $(\bar{x})$ y $(\bar{y})$ son las medias de $(X)$ y $(Y)$.
- $(n)$ es el número total de observaciones.

Se dice que dos variables tienen una relación fuerte si:

$$|r(X,Y)| \geq \frac{2}{\sqrt{n}}$$

### Propiedades del coeficente de correlación de Pearson

1. $$r(X,Y)=r(Y,X)$$
2. $$r(X,Y)=r(aX + b,Y)\text{, con a y b constantes.}$$
3. $$|r(X,Y)|\leq 1$$
4. $$|r(X,Y)|=1\text{, Si }y_{i}=ax_{i} + b$$

### Limitaciones

- Sensibilidad a outliers: Valores atípicos pueden distorsionar el coeficiente y dar lugar a conclusiones erróneas.

- Asume una relación lineal: Si las variables están relacionadas de manera no lineal, el coeficiente de Pearson puede ser inadecuado para describir dicha relación.

- No detecta relaciones no lineales: Un coeficiente cercano a 0 no significa que no exista relación, sino que no hay relación lineal.


# Diagrama de dispersión y asociación de variables <a id="scatterplot_simulation"></a>

Como ya hemos visto, el diagrma de dispersión es útil para describir dos variables cuantitativas. En este apartado vamos a visualizar dos variables $X$ e $Y$ de tal forma que $Y = f(X) + \epsilon$. Con ello calculamos el coeficiente de Pearson y analizamos el comportamiento de dicho coeficiente en diferentes situaciones.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


############
## random ##
############
np.random.seed(42)
x = np.random.uniform(0, 1, 100)
##############################
## Relación lineal positiva ##
##############################
a1 = 3.6
b1 = 5
y1 = a1 * x + b1 + np.random.normal(0, 0.3, len(x))

##############################
## Relación lineal negativa ##
##############################
a2 = -2
b2 = 3
y2 = a2 * x + b2 + np.random.normal(0, 0.3, len(x))

#########################
## Relación cuadrática ##
#########################
a3 = 10
b3 = 2
y3 = a3 * x**2 + b3 + np.random.normal(0, 0.5, len(x))

##########################
## Relación logarítmica ##
##########################
a4 = 2
b4 = 1
y4 = a4 * np.log(x) + b4 + np.random.normal(0, 0.5, len(x))

##########################
## Relación exponencial ##
##########################
a5 = 10
b5 = 1
y5 = a5 * np.exp(2*x) + b5 + np.random.normal(0, 1, len(x))

#########################
## Relación sinusoidal ##
#########################
a6 = 10
b6 = 1
y6 = a6 * np.sin(10*x) + b6 + np.random.normal(0, 0.5, len(x))

#################
## No relación ##
#################

y7 = np.random.normal(10, 1, len(x))
```

```python
################
## Data Frame ##
################

df = pd.DataFrame(
    data={
        "x": x,
        "lineal_positiva": y1,
        "lineal_negativa": y2,
        "cuadratica": y3,
        "logaritmica": y4,
        "exponencial": y5,
        "sinusoidal": y6,
        "no_relacion": y7
    }
)

cols_list = [col for col in df.columns.to_list() if col != "x"]
cols_subplot =  2
rows_suplot = 4
fig, axs = plt.subplots(rows_suplot, cols_subplot, figsize=(8, 10), constrained_layout=True)
axs = axs.ravel()


for index, col in enumerate(cols_list):
    pearson_corr = np.corrcoef(df["x"], df[col])[0, 1]
    axs[index].scatter(df["x"], df[col])
    axs[index].set_title(col)
    axs[index].set_xlabel("x")
    axs[index].set_ylabel(col)
    axs[index].text(0.8, df[col].min(), f"r = {pearson_corr:.2f}", fontsize=12, color='blue')
```
