# **Conceptos básicos** <a id="basic_concepts"></a>

<br>

## **Experimento aleatorio** <a id="random_experiment"></a>

- **Experimento**: Es un proceso bien definido y reproducible que puede ser realizado bajo las *mismas* condiciones

- **Experimento aleatorio**: Es un experimento del cual su resultado no puede ser determinado con anterioridad. De un experimento aleatorio se conocen todos sus *posibles resultados*.

<br>

## **Espacio muestral** <a id="sample_space"></a>

El espacio muestral de un experimento aleatorio es el conjunto de todos sus *posibles resultados*, se nota con la letra $\Omega$.

Los elementos que conforman el espacio muestral $(\Omega)$ se denominan puntos muestrales y se notan con la letra $\omega$.

Cuando el $|\Omega|$ es finito o numerable se dice que el experimento aleatorio es discreto.

<br>

- **Ejemplos**:

    - Se observa el número de veces que es necesario lanzar una moneda corriente hasta obtener cara por primera vez
$$\Omega := \{ 1,2,3,\dots , \infty \}$$

    - Se observa el nivel del embalse de agua potable de Chingaza
$$\Omega := \{x : x \in [a,b]\}$$

<br>

## **Sigma álgebra** <a id="sigma_field"></a>

De manera natural surge la pregunta *¿Cuál es la chance que tiene un "evento" de ocurrir?*.

- ¿Qué es un evento?

Un evento es un subconjunto del espacio muestral que debe pertenecer a una estructura llamada $\sigma-\text{álgebra}$. Con esto queda claro que **NO** todo subconjunto del espacio muestral $\Omega$ es un evento

<br>

### **Definción**


Sea $\Omega \neq \varnothing$. Una colección $\mathcal{F}$ de subconjuntos de $\Omega$ es una $\sigma\text{-álgebra}$ sobre $\Omega$, si:


$$
\begin{array}{ll}
\textit{i)} & \Omega \in \mathcal{F}.\\
\textit{ii)} & \text{Si } A \in \mathcal{F} \text{ entonces } A^c \in \mathcal{F}.\\
\textit{iii)} & \text{Si } A_1, A_2, \dots \in \mathcal{F} \text{ entonces } \bigcup_{i=1}^\infty A_i \in \mathcal{F}.
\end{array}
$$

**Los elementos de $\mathcal{F}$ se llaman eventos**.

<br>

- **Ejemplos**

    a. Si $\Omega \neq \varnothing$ entonces $\mathcal{F} := \lbrace \varnothing, \Omega \rbrace$, se denomina $\sigma$-álgebra trivial.

    b. Si $\Omega \neq \varnothing$ entonces $\mathcal{F} := \lbrace A : A \subseteq \Omega \rbrace$, se denomina $\sigma$-álgebra total, se nota $2^{\Omega}$.


<br>

- **Ejercicio**

    b. Construya una $\sigma\text{-álgebra}$  que *NO* sea la total *NI* la trivial con el experimento aleatorio: Lanzar dos monedas.


## **Espacio medible** <a id="measurable_space"></a>

Sean $\Omega \neq \varnothing$ y $\mathcal{F}$ una $\sigma-\text{álgebra}$ sobre $\Omega$. Entonces la dupla $(\Omega, \mathcal{F})$ se le llama espacio medible.

- Por definición $\Omega, \varnothing \in \mathcal{F}$
- Un evento $\lbrace \omega \rbrace$, se denomina evento elemental.
- Un evento $A \in \mathcal{F}$ ocurre, si el resultado del experimento $\omega \in A$.

<br>

Si $A$ y $B$ son dos eventos se tiene que:

$$
\begin{array}{ll}
\textit{i)} & A \cup B \text{ es un evento que ocurre, si y sólo si, } A \text{ o } B \text{ o ambos ocurren.} \\[8pt]
\textit{ii)} & A \cap B \text{ es un evento que ocurre, si y sólo si, } A \text{ y } B \text{ ocurren.} \\[8pt]
\textit{iii)} & A^c \text{ es un evento que ocurre, si y sólo si, } A \text{ no ocurre.} \\[8pt]
\textit{iv)} & A - B \text{ es un evento que ocurre, si y sólo si, } A \text{ ocurre pero } B \text{ no.}
\end{array}
$$

<br>

**Ejemplo**

El experimento aleatorio es el lanzamiento de un dado corriente 3 veces, por lo tanto:

$$\Omega := \lbrace (a,b,c): a,b,c \in \lbrace 1,2,3,4,5,6 \rbrace \rbrace$$

<br>

- $A$: El resultado del primer lanzamiento es un número primo.
- $B$: La suma de los resultados obtenidos es menor o igual a 4.

<br>

$$
A \cup B = \{ (a, b, c) \in \Omega : a \in \{2, 3, 5\} \text{ o } (a + b + c) \leq 4 \}
$$

$$
A \cap B = \{ (a, b, c) : a \in \{2, 3, 5\} \text{ y } (a + b + c) \leq 4 \} = \{(2, 1, 1)\}
$$

<br>

**Ejercicio**

Con el experimento anterior definir a que es igual:
- $A-B$
- $A^c$

## **Eventos mutuamente excluyentes** <a id="excluding_events"></a>

Sean dos eventos $A$ y $B$. Se dice que son mutuamente excluyentes si:

$$A \cap B = \varnothing$$

<br>

**Ejemplo**

El experimento es el lanzamiento de una moneda.

<br>

- $C$: El resultado es cara
- $S$: El resultado es sello

Evidentemente son eventos mutuamente excluyentes.

## **Concepto de probabilidad** <a id="probability_concept"></a>

El objetivo es asignar a cada evento $A$ un número real no negativo que indique la "chance" que tiene de ocurrir.

$$f: \Omega \longrightarrow [0,1]$$

<br>

## **Frecuencia relativa** <a id="relative_frequency"></a>

Suponga que se realiza el experimento aleatorio $n$ veces y que las condiciones se mantienen relativamente constantes, entonces la frecuencia relativa se define así:

$$f_{r}(A):= \frac{n(A)}{n}$$

Donde $n(A)$ es el número de veces que ocurre $A$.

<br>

*Consideraciones*

- El número $f_{r}(A)$ no es constante, depende del número de veces que se realiza el experimento.

- Cuando $n >>>$, el número $f_{r}(A)$ tiende a estabilizarce entre $0$ y $1$.

<br>

**Ejemplo**
<br>

Se lanza un dado corriente n veces y se define un evento así:

$$A:= \text{El resultado obtenido es 3.}$$

<br>

| \(n\)  | Frecuencia | Frecuencia Relativa |
|--------|------------|---------------------|
| 100    | 14         | 0.14               |
| 200    | 29         | 0.145              |
| 300    | 51         | 0.17               |
| 400    | 65         | 0.1625             |
| 500    | 83         | 0.166              |


<br>


***Consideraciones***

El valor sobre el que $f_{r}(A)$ tiende a estabilizarce se denomina $P(A)$, con esto se tiene que:

- Dado que $n(A) \geq 0$, entonces $P(A) \geq 0$.
- Dado que $n(\Omega) = n$, entonces $P(\Omega) = 1$.

<br>

Además, si $A \cap B = \varnothing$, entonces:

- $n(A \cap B) = 0$.
- $n(A \cup B) = n(A) + n(B)$.
- $f_{r}(A \cup B) = f_{r}(A) + f_{r}(B)$
- $P(A \cup B) = P(A) + P(B)$


Con la anterior quedan sentadas las base para definir un espacio de probabilidad.

<br>

## **Espacio de probabilidad** <a id="probability_space"></a>

Sea $(\Omega, \mathcal{F})$ un espacio medible. Una función $(P)$ definida sobre $\mathcal{F}$ y de valor real que satisface las siguientes condiciones:

1. $P(A) \geq 0$ para todo $A \in \mathcal{F}$.  
2. $P(\Omega) = 1$.  
3. Si $A_1, A_2, \dots$ son elementos de $\mathcal{F}$ mutuamente excluyentes, esto es  
$$
   A_i \cap A_j = \varnothing \quad \text{para todo } i \neq j,
$$
   entonces  
$$
   P\left( \bigcup_{i=1}^\infty A_i \right) = \sum_{i=1}^\infty P(A_i).
$$

Se llama medida de probabilidad sobre $(\Omega, \mathcal{F})$.  
La tripla $(\Omega, \mathcal{F}, P)$ se llama espacio de probabilidad.

<br>


**Ejemplo**  
Sean $\Omega = \{1, 2\}$, $\mathcal{F} = 2^{\Omega}$ y $P$ la aplicación definida sobre $\mathcal{F}$ por:

$$
P(A) =
\begin{cases}
0 & \text{si } A = \varnothing, \\
\frac{1}{3} & \text{si } A = \{1\}, \\
\frac{2}{3} & \text{si } A = \{2\}, \\
1 & \text{si } A = \{1, 2\}.
\end{cases}
$$

$P$ es una medida de probabilidad.


<br>

## **Evento nulo** <a id="null_event"></a>

Un espacio de probabilidad $(\Omega, \mathcal{F}, P)$, cualquier evento $A$ tal que $P(A) = 0$, se llama evento nulo

<br>

## **Espacio completo** <a id="complete_space"></a>

Un espacio de probabilidad $(\Omega, \mathcal{F}, P)$, se dice completo si todos los subconjuntos de eventos nulos son también eventos nulos.

<br>

**Ejemplo**

Sea $\Omega = \{1, 2, 3\}$, $\mathcal{F} = \{\varnothing, \{2\}, \{1, 3\}, \Omega\}$ y $P$ dada por:

$$
P(A) =
\begin{cases}
1 & \text{si } 2 \in A, \\
0 & \text{si } 2 \notin A.
\end{cases}
$$

Los eventos $\varnothing$ y $\{1, 3\}$ son eventos nulos.

## **Propiedades de una medida de probabilidad** <a id="probability_measure_properties"></a>

**Teorema**

Sea $(\Omega, \mathcal{F}, P)$ un espacio de probabilidad. Entonces:

1. $P(\varnothing) = 0$.  

2. Si $A, B \in \mathcal{F}$ y $A \cap B = \varnothing$, entonces  
   $$
   P(A \cup B) = P(A) + P(B).
   $$

3. $P(A^c) = 1 - P(A)$.  

4. Si $A \subseteq B$, entonces $P(A) \leq P(B)$ y $P(B - A) = P(B) - P(A)$. En particular, se tiene que $P(A) \leq 1$ para todo $A \in \mathcal{F}$.

5.  
   $$
   P(A \cup B) = P(A) + P(B) - P(A \cap B).
   $$

6. Sea $(A_n)$ una sucesión creciente de elementos de $\mathcal{F}$, esto es $A_n \in \mathcal{F}$ para todo $n$ y además $A_n \subseteq A_{n+1}$ para todo $n = 1, 2, \dots$, entonces:  
   $$
   P\left(\lim_{n \to \infty} A_n\right) = \lim_{n \to \infty} P(A_n),
   $$
   donde  
   $$
   \lim_{n \to \infty} A_n = \bigcup_{n=1}^\infty A_n.
   $$

7. Sea $(A_n)$ una sucesión decreciente de elementos de $\mathcal{F}$, esto es $A_n \in \mathcal{F}$ para todo $n$ y además $A_n \supseteq A_{n+1}$ para todo $n = 1, 2, \dots$, entonces:  
   $$
   P\left(\lim_{n \to \infty} A_n\right) = \lim_{n \to \infty} P(A_n),
   $$
   donde  
   $$
   \lim_{n \to \infty} A_n = \bigcap_{n=1}^\infty A_n.
   $$


<br>

*Las demostraciones se pueden ver en Blanco Castañeda, L. (2004).*

<br>


***Consideraciones***

Sean $(\Omega, \mathcal{F}, P)$ un espacio de probabilidad con $\Omega$ finito o numerable y $\mathcal{F} = 2^{\Omega}$. Sea $\varnothing \neq A \in \mathcal{F}$. Es claro que  

$$
A = \bigcup_{\omega \in A} \{\omega\}
$$

y por consiguiente  

$$
P(A) = \sum_{\omega \in A} P(\omega),
$$

donde $P(\omega) := P(\{\omega\})$. Esto es, $P$ está completamente determinada por $p_j := P(\omega_j)$, donde los $\omega_j$, $j = 1, 2, \dots$, denotan los elementos de $\Omega$.  

Es claro que el vector $p := (p_1, p_2, \dots)$ de dimensión $|\Omega|$, donde $|\Omega|$ denota el número de elementos de $\Omega$, satisface las condiciones:

1. $$p_j \geq 0$$

2.  
   $$
   \sum_{j=1}^\infty p_j = 1
   $$

Un vector $p$ que satisface las condiciones anteriores se llama vector de probabilidades.

<br>

## **Espacio laplaciano de probabilidad** <a id="laplacian_space"></a>

Un espacio de probabilidad $(\Omega, \mathcal{F}, P)$ se llama laplaciano, si $\Omega$ es finito, $\mathcal{F} = 2^{\Omega}$ y $P(\omega) = \frac{1}{|\Omega|}$ para todo $\omega \in \Omega$. La medida de probabilidad $P$ se llama **distribución laplaciana** (o uniforme o clásica) en $\Omega$.

**Nota**  
Si $(\Omega, \mathcal{F}, P)$ es un espacio de probabilidad laplaciano y $A \subseteq \Omega$, entonces:

$$
P(A) = P\left( \bigcup_{\omega \in A} \{\omega\} \right) = \sum_{\omega \in A} \frac{1}{|\Omega|} = \frac{|A|}{|\Omega|}.
$$

En otras palabras,

$$
P(A) = \frac{\text{"número de casos favorables a } A\text{"}}{\text{"número de casos posibles"}}.
$$

*Esta última expresión no es de ningún modo una definición de probabilidad, sino simplemente una consecuencia de suponer que todos los resultados son igualmente probables*.

<br>

**Ejemplo**

En un taller de reparación de electrodomesticos se encuentran 10 televisores para reparación

- Marca $A=3$
- Marca $B=3$
- Marca $C=4$

<br>

1. Cuál es la probabilidad $p_1$ de que el primer televisor reparado sea de la marca A:

    - Es claro que existen $10!$ formas diferentes de reparar los televisores.
    - Para reparar en primer lugar un televisor de la marca $A$, se puede escoger primero el televisor $A_1$ o el $A_2$ o el $A_3$ entonces:

        $$P(A_1) + P(A_2) + P(A_3)$$

        $$P(A_i) = \frac{9!}{10!}$$


En consecuencia se tiene que $3P(A_i)=p_1=\frac{3 \times 9!}{10!}$

<br>

2. Cuál es la probabilidad $p_2$ de que los tres primeros televisores sean de la marca A:

    - En este caso los tres primeros elementos son de la marca A, estos se pueden permutar entre ellos de $3!$ formas.

        $$\frac{3!7!}{10!}$$


<br>

3. La probabilidad $p_3$ de que el orden de reparación de los televisores sea (CABCABCABC)

    - Acá hacemos uso del siguiente razonamiento: Como existen 3 tipos de elementos que se consideran indistinguibles o iguales (marcas de telivisor), lo que hacemos es tomar el orden dado (CABCABCABC) y calcular de cuantas formas pueden permutarse cada una de los elementos indistinguibles (Marcas) en el orden requerido (CABCABCABC). Para la marca A hay $3!$ formas, para la marca B $3!$ y para la marca C $4!$, en consecuencia se tiene:
    
    $$\frac{4!3!3!}{10!}$$

<br>
