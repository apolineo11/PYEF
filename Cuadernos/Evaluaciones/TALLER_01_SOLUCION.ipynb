La variabilidad relativa con respecto a la media $(CV)$ nos indica que el tratamiento produce datos con casi el doble de variablilidad relativa $(12\%)$, mientras que en los pacientes sin recibir dicho tratamiento teníamos un $CV$ de $(7\%)$

<br>


|Estadístico | Antes | Después |
|------------|:-----:|:-------:|
|$\bar{x}$   |134.20 |121.50   |

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from scipy.stats import spearmanr, kendalltau

# data
before = np.array([144, 124, 148, 116, 132, 129, 141, 134, 139, 135])
after = np.array([138, 120, 131, 95, 109, 133, 112, 117, 115, 145])


# before
x_bar_before = np.mean(before)
print("----")
print(f"El promedio antes del tratamiento es {x_bar_before:.2f}")

sigma_before = np.std(before, ddof=1)
print("----")
print(f"La desviación estándar antes del tratamiento es {sigma_before:.2f}")


variation_coef_before = sigma_before / x_bar_before
print("----")
print(f"El coeficientes de variación antes del tratamiento es {variation_coef_before:.2f}")

# after
x_bar_after = np.mean(after)
print("----")
print(f"El promedio después del tratamiento es {x_bar_after:.2f}")

sigma_after = np.std(after, ddof=1)

print("----")
print(f"La desviación estándar después del tratamiento e {sigma_after:.2f}")


variation_coef_after = sigma_after / x_bar_after

print("----")
print(f"El coeficientes de variación después del tratamiento es {variation_coef_after:.2f}")



```

a. Se lanzan al mismo tiempo 4 dados indistinguibles ¿Cuántos resultados diferentes pueden obtenerse?

Si los dados fueran distinguibles, cada dado de estos puede tener $6$ resultados diferentes, entonces tendriamos que son posibles $6 \times 6 \times 6 \times 6 = 6^{4}$.

Sin embargo, en dados indistinguibles es igual sacar $(2,6,6,5)$ que $(6,5,2,6)$. Con esto, el problema puede ser visto como el número de soluciones no negativas de la siguiente ecuación:

$$x_{1} +x_{2} + x_{3} + x_{4} + x_{5} + x_{6} = 4$$

En donde $x_{i}$ = Nuḿero de veces que sale el valor $i$ como resultado. $i \in \lbrace 1,2,3,4,5,6 \rbrace$.

Con esto el numero de resultados posibles es:

$$\binom{4+6-1}{6-1}=\frac{9!}{4!5!}=126$$

b. Tres atletas participan en una competencia ¿De cuantas maneras pueden llegar a la meta si existe la posibilidad de que lleguen juntos?


Como pueden llegar juntos se deben considerar 3 situaciones:

- Cuando no llegan juntos

$$P(3,3)=3!$$

- Cuando llegan 2 personas juntas
$$C(3,2)\times P(2,2) = 3!$$

- Cuando los 3 llegan juntos

$$C(3,3) = 1$$

Entonces se tienen $13$ formas de llegar a la meta si pueden llegar juntos

c. Realice una tabla de frecuencias (absoluta, relativa, absoluta acumulada, relativa acumulada) de la cantidad de personas en los hogares por cada nivel educativo y haga un breve análisis.

```python
import pandas as pd
from copy import deepcopy

population_df = pd.read_excel("https://raw.githubusercontent.com/apolineo11/PYEF/main/Cuadernos/Data/hogares.xlsx")
population_df['Nivel Educativo'] = pd.Categorical(population_df['Nivel Educativo'], categories=['Sin educación formal', 'Primaria', 'Secundaria', 'Universitario'], ordered=True)
population_df['Región'] = pd.Categorical(population_df['Región'], categories=['Rural', 'Urbano'], ordered=True)
df_list  = [deepcopy(population_df[population_df['ID'].isin(idx_k)]) for idx_k in idxs]
```

```python
columns_to_normalize = ['Sin educación formal',	'Primaria',	'Secundaria', 'Universitario']

for k, df in enumerate(df_list):
    educational_level = df.loc[:,['Nivel Educativo', 'Tamaño del Hogar']].groupby(['Nivel Educativo', 'Tamaño del Hogar'], observed=False).size().reset_index(name='Frecuencia Absoluta')
    frequency_abs = educational_level.pivot(index='Tamaño del Hogar', columns='Nivel Educativo', values='Frecuencia Absoluta').reset_index().fillna(0).rename_axis(None, axis=1)
    frequency_rel = frequency_abs[columns_to_normalize].div(frequency_abs[columns_to_normalize].sum(axis=0), axis=1).fillna(0)
    frequency_rel = pd.concat([frequency_abs['Tamaño del Hogar'], frequency_rel], axis=1)
    frequency_abs_cum = frequency_abs[columns_to_normalize].cumsum()
    frequency_abs_cum = pd.concat([frequency_abs['Tamaño del Hogar'], frequency_abs_cum], axis=1)
    frequency_rel_cum = frequency_rel[columns_to_normalize].cumsum()
    frequency_rel_cum = pd.concat([frequency_rel['Tamaño del Hogar'], frequency_rel_cum], axis=1)
    print("----")
    print(f"Para el K = {k_array[k]}")
    print("Tabla de frecuencias absolutas")
    display(frequency_abs)
    print("Tabla de frecuencias relativas")
    display(frequency_rel)
    print("Tabla de frecuencias absolutas acumuladas")
    display(frequency_abs_cum)
    print("Tabla de frecuencias relativas acumuladas")
    display(frequency_rel_cum)
    print("----")
```

e. Describir y analizar la distribución de la variable ingreso mensual. Para determinar las clases use la regla de Sturges

```python
bins_sturges = int(1 + np.log2(20))
bins_rice = int(2*20**(1/3))
print(f"El número de clases para la regla de Sturges es {bins_sturges}")
print(f"El número de clases para la regla de Rice es {bins_rice}")
```

```python
fig, axs = plt.subplots(1, len(k_array), figsize=(20,5))

for k, df in enumerate(df_list):
    median_value = df['Ingreso Mensual (USD)'].median()
    axs[k].hist(df['Ingreso Mensual (USD)'], bins=bins_sturges, alpha=0.6, density=False)
    axs[k].axvline(median_value, color='green', linestyle='--', label=f'Median: {median_value:.2f}')


    # density kernel

    x_vals = np.linspace(min(df['Ingreso Mensual (USD)']), max(df['Ingreso Mensual (USD)']), 500)
    bin_edges = np.linspace(min(df['Ingreso Mensual (USD)']), max(df['Ingreso Mensual (USD)']), bins_sturges + 1)
    bin_width = bin_edges[1] - bin_edges[0]
    kde = gaussian_kde(df['Ingreso Mensual (USD)'])
    density_vals = kde(x_vals)*bin_width*len(df['Ingreso Mensual (USD)'])
    axs[k].plot(x_vals, density_vals, color='red', label='KDE Curve')

    axs[k].set_title(f"Histograma del Ingreso Mensual (USD)\nRegla de Sturges\nMuestreo (K) = {k_array[k]}")
    axs[k].set_xlabel('Ingreso Mensual (USD)')
    axs[k].set_ylabel('Frecuencia')
    axs[k].legend()
    axs[k].set_ylim(0, 10)
fig.show()
```
