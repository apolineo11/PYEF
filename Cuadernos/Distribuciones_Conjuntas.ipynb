# **Variables Aleatorias Distribuidas Conjuntamente**

## **1. Introducción**
En teoría de la probabilidad, a menudo trabajamos con múltiples variables aleatorias que están relacionadas. El concepto de **variables aleatorias distribuidas conjuntamente** nos ayuda a entender cómo interactúan dos o más variables aleatorias.

Por ejemplo, consideremos un experimento donde medimos la altura y el peso de individuos. Estas dos variables no son independientes, y su relación puede describirse utilizando distribuciones de probabilidad conjunta.

---

## **2. Distribuciones de Probabilidad Conjunta**
### **2.1 Función de Distribución Acumulativa Conjunta (CDF)**
Para dos variables aleatorias $X$ y $Y$, la **función de distribución acumulativa conjunta** se define como:

$$
F(x, y) = P(X \leq x, Y \leq y)
$$

Esta función da la probabilidad de que $X$ sea menor o igual a $x$ y $Y$ sea menor o igual a $y$.

---

### **2.2 Función de Masa de Probabilidad Conjunta (PMF) para Variables Discretas**
Si $X$ y $Y$ son variables aleatorias discretas, la **función de masa de probabilidad conjunta** (PMF) se define como:

$$
p(x, y) = P(X = x, Y = y)
$$

Las probabilidades marginales pueden obtenerse sumando sobre los valores posibles de la otra variable:

$$
P(X = x) = \sum_{y} p(x, y), \quad P(Y = y) = \sum_{x} p(x, y)
$$

#### **Ejemplo: Cálculo de la PMF Conjunta**
```python
import numpy as np
import pandas as pd

# Ejemplo de tabla de PMF conjunta
data = np.array([[0.1, 0.2, 0.1],
                 [0.15, 0.2, 0.25]])

df = pd.DataFrame(data, index=["X=0", "X=1"], columns=["Y=0", "Y=1", "Y=2"])
df
```

---

### **2.3 Función de Densidad de Probabilidad Conjunta (PDF) para Variables Continuas**
Si $X$ y $Y$ son variables aleatorias continuas, definimos la **función de densidad de probabilidad conjunta** (PDF) como:

$$
P(a \leq X \leq b, c \leq Y \leq d) = \int_{c}^{d} \int_{a}^{b} f(x, y) \ dx \ dy
$$

Las densidades marginales están dadas por:

$$
f_X(x) = \int_{-\infty}^{\infty} f(x, y) \ dy, \quad f_Y(y) = \int_{-\infty}^{\infty} f(x, y) \ dx
$$

#### **Ejemplo: Cálculo de la PDF Conjunta**
```python
from sympy import symbols, exp, integrate

x, y = symbols('x y')

# Ejemplo de PDF f(x,y) = 2e^(-x) e^(-2y)
f_xy = 2 * exp(-x) * exp(-2*y)

# Calcular P(X > 1, Y < 1)
prob = integrate(integrate(f_xy, (x, 1, float('inf'))), (y, 0, 1))
prob.evalf()
```

## **3. Independencia de Variables Aleatorias**
Dos variables aleatorias $X$ y $Y$ son **independientes** si y solo si su función de probabilidad conjunta se factoriza:

$$
F(x, y) = F_X(x) F_Y(y)
$$

Para variables discretas:

$$
p(x, y) = p_X(x) p_Y(y)
$$

Para variables continuas:

$$
f(x, y) = f_X(x) f_Y(y)
$$

#### **Ejemplo: Verificación de Independencia**
```python
# Distribución conjunta dada
pxy = np.array([[0.25, 0.25], [0.25, 0.25]])
px = pxy.sum(axis=1)
py = pxy.sum(axis=0)

# Verificar independencia
independiente = np.allclose(pxy, np.outer(px, py))
independiente
```

---

## **4. Distribuciones Condicionales**
La **función de masa de probabilidad condicional** para variables discretas es:

$$
P(X = x | Y = y) = \frac{P(X = x, Y = y)}{P(Y = y)}
$$

Para variables continuas, la **función de densidad condicional** es:

$$
f_{X|Y}(x | y) = \frac{f(x, y)}{f_Y(y)}
$$

#### **Ejemplo: Cálculo de Probabilidad Condicional**
```python
# PMF conjunta dada
pxy = np.array([[0.2, 0.3], [0.1, 0.4]])
py = pxy.sum(axis=0)

# Probabilidad condicional P(X | Y)
p_x_given_y = pxy / py
p_x_given_y
```

---

## **5. Resumen**
- **Las distribuciones conjuntas** describen la relación entre múltiples variables aleatorias.
- **Las distribuciones marginales** se obtienen sumando o integrando sobre la otra variable.
- **La independencia** ocurre cuando la distribución conjunta se factoriza.
- **Las distribuciones condicionales** ayudan a comprender el comportamiento de una variable dado el valor de otra.
