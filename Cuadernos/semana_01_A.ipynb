# ¿QUÉ ES LA ESTADÍSTICA?

- La estadística tiene pues por objeto presentar una representación fiel de un estado en una época determinada. (Quetelet, 1849)

- La estadística es el único instrumento que permite abrirse paso a través de la formidable espesura de dificultades que obstaculiza el camino de los que persiguen la Ciencia del hombre. (Galton, 1889)

- La estadística puede considerarse (i) como el estudio de las poblaciones, (ii) como el estudio de la variación, y (iii) como el estudio de los métodos de reducción de datos. (Fisher, 1925)

- La estadística es una disciplina científica que se ocupa de la recogida, el análisis y la interpretación de los datos obtenidos a partir de la observación o la experimentación. La estadística tiene una estructura coherente basada en la teoría de la Probabilidad e incluye muchos procedimientos diferentes que contribuyen a la investigación y el desarrollo en toda la Ciencia y la Tecnología. (E. Pearson, 1936)

- Estadística es el nombre de la ciencia y el arte que se ocupa de las inferencias inciertas. Que utiliza números para averiguar algo sobre la naturaleza y la experiencia. (Weaver, 1952)

- En el siglo XX, la estadística se conoce como la herramienta matemática para analizar datos experimentales y observacionales. (Porter, 1986)

- La estadística es el arte de aprender de los datos. (Ross, 2020)


## ¿Y entonces qué?

1. **Variabilidad**:
    - Reconocimiento de que los datos tienden a distribuirse de forma diferente y que tal comportamiento depende de  muchos factores.

2. **Incertidumbre**:
     - Toma de decisiones en contextos de incertidumbre.

3. **Descriptiva**:
    - Herramienta para describir estados o realidades de interés.

4. **Inferencial**:
    - Poder realizar afirmaciones sobre las características de una población.



La estadística es **experimental y observacional**. Pero tiene su gran soporte en la teoria de probabilidad y en sus propios desarrollos metodológicos. En este sentido su matería prima de trabajo son los **datos**.


# Población y Muestra

## Población

La **POBLACIÓN** es la totalidad de elementos que componen el conjunto que es de interés. Esta colección de objetos debe ser bien definida, no cabe lugar a la ambigüedad. Por lo general se hace de forma similar a la definición de un conjunto. El total de la población se nota con la letra $N$.

**Ejemplos**:

- Una investigadora le interesa estudiar el rendimiento académico de los estudiantes de la asignatura Probabilidad y Estadística fundamental. Entonces su población de estudio puede ser definidad como *TODOS* los estudiantes de la universidad que tienen inscrita dicha asignatura en el presente semestre.

- A un apostador le interesa comprender el mercado de jugadores de los equipos europeos de los últimos 10 años. La población son todos los fichajes de jugadores en los que intervenga un equipo de Europa.


## Muestra

En muchas ocasiones, y a pesar de la era del BIG DATA, muchas veces no es posible registrar los datos de todos los elementos de la población. El subconjunto de elementos de la población del que tenemos registro se denomina **MUESTRA**.

**Ejemplos**

- La investigadora que quiere estudiar el rendimiento académico de los estudiantes de la asignatura Probabilidad y Estadística fundamental no puede tener acceso a los registros de todos los estudiantes y unicamente tiene acceso a los grupos 5 y 7. Entonces, la muestra son los estudiantes de los grupos 5 y 7.

- En la Europa oriental no existen registros de los fichajes de los últimos diez años, entonces se ve obligado a tomar el registro de los paises restantes. Y esta es su muestra.

# Muestreo

La forma en la que se elijen o seleccionan los elementos de la muestra se denomina muestreo. Los principales retos del muestro son:

- Determinar el tamaño adecuado de la muestra $(n)$.
- Seleccionar el método de muestreo adecuado.

El muestreo es importante ya que con base en la muestra se busca hacer afirmaciones de la población.

## Métodos de muestreo (Hay otros)

1. **Muestreo aleatorio simple:**
    - Todos los miembros de la población tienen la misma probabilidad de ser elegidos.
    - La elementos son seleccionados de forma independiente. La elección de uno NO condiciona la elección de otro(s).
    - Se usa el "azar", generador de números pseudoaletorios.

2. **Muestreo Sistemático**:
    - "Ordenar" la población
    - Definir el tamaño de muestra $n$
    - Variable $j = \frac{N}{n}$
    - Escoger un valor $1 \leq k \leq j$
    - El elemento ubicado en la posicion $k$ es el primer elemento de la muestra
    - El restante de elementos se selecciona de manera recurrente sumando el valor j al valor previo obtenido:

Ejemplo: Se tiene una población de tamaño $100$ y se desea una muestra de tamaño $11$.
$$j = 9.09 ≈ 9$$
$$k = 2$$
$$s_{1} = 2$$
$$s_{2} = s_{1} + j$$
$$s_{3} = s_{2} + j$$


## Unidad experimental

Es un individuo o elemento de la muestra en el que se miden o registran cosas.

 # Parámetro y estadístico

1. **Parámetro**: Un parámetro es una medida numérica que describe una característica específica de una población. Son una constantes que representan la población.

2. **Estadístico**: Un estadístico es una medida numérica que describe una característica específica de una muestra. Es una función que no depende de constantes desconocidas ni de los parámetros de la población.


## Estadística Inferencial y Estadística Descriptiva

- **Estadística Descriptiva**: Cuando el ejercicio hecho tiene como objetivo reconocer,caracterizar y resumir caracteristicas importantes de un conjunto de medidas un fenómeno se hace estadística descriptiva. Para ello se utilizan herramientas graficas y resúmenes númericos.

- **Estadística Inferencial**: Si el propósito es realizar afirmaciones, predicciones y hacer decisiones sobre una población a partir de una muestra.

# Proceso de análisis de inferencia estadística

Antes de iniciar cualquier proceso de análisis de tipo inferencial es fundamental **Definir el objeto de estudio** y **Determinar que información se necesita**, luego podemos iniciar así:

1. Especificar las preguntas a ser respondidas e identificar la población de interés.

2. Decidir el *como* seleccionar la muestra

3. Seleccionar la muestra y analizar la información de esta.

4. Usar la información obtenida en $3$ para hacer inferencia sobre la población.

5. Determinar la fiabildad de las inferencias hechas.


# Variables

- **Variable**: Es una característica que cambia en el tiempo y/o en cada uno de los individuos u objetos de interés.

- **Variable cuantitativa**: Medida numérica de una cantidad sobre cada unidad experimental.
    - **Variable discreta**: Solo puede tomar un numero finito o contable de valores.
    - **Variable continua**: Puede asumir los infinitos valores correspondientes a los puntos de un intervalo de rectas.

- **Variable cualitativa**: Miden una cualidad o característica de cada unidad experimental. Las variables cualitativas producen datos que pueden ser separados en categorías y por eso usualmente se llaman variables **categóricas**.

# Escalas de medida

Esta es una división de las varaibles que se da en función del **SENTIDO** o **NO** que tienen ciertas operaciones matemáticas que se pueden hacer entre ellas.

A. **Nominales**:
   En este nivel solo se pueden tener relaciones de igualdad entre los valores de la variables. Con tales variables se pueden realizar conteos y agrupaciones.

B. **Ordinales**: Indican el orden que ocupan los objetos y, al igual que en el caso de los datos nominales, los valores son palabras que describen las respuestas

C. **Intervalo**: Incluye las características de una variable con escala de medida ordinal, pero adicionalmente tiene sentido tanto la suma como la resta de valores (las distancias entre los valores tienen sentido).

D. **Razón**: Incluye las características de una variable con escala de medida de intervalo, pero adicionalmente tienen sentido la multiplicación y división de valores de la variable (la razón entre los valores tiene sentido) y el valor cero representa la ausencia de lo que mide la variable (el cero es significativo).
