# Sección 1: Variables Aleatorias Reales
# =================================================

## Definición y conceptos básicos

Una variable aleatoria es una función que asigna a cada elemento del espacio muestral un número real, formalizando así los resultados de un experimento aleatorio en términos matemáticos. La idea detrás de este concepto es representar cuantitativamente los eventos del experimento y facilitar el análisis probabilístico

Existen dos tipos principales de variables aleatorias reales: discretas y continuas. Una variable aleatoria discreta toma valores específicos y contables, mientras que una variable aleatoria continua puede tomar cualquier valor dentro de un intervalo.

Sea un espacio de probabilidad $(\Omega, \mathcal{F}, P)$, donde $\Omega$ es el espacio muestral, $\mathcal{F}$ una $\sigma$-álgebra de $\Omega$ y $P$ una medida de probabilidad.

Una variable aleatoria $X$ es una función medible $X: \Omega \to \mathbb{R}$ que satisface que, para cualquier conjunto Boreliano $B \subseteq \mathbb{R}$, el conjunto $X^{-1}(B) = \{\omega \in \Omega : X(\omega) \in B\}$ pertenece a $\mathcal{F}$.

Por ejemplo, en un experimento de lanzar un dado, el espacio muestral es $\Omega = \{1, 2, 3, 4, 5, 6\}$. Definimos una variable aleatoria $X$ que asigna a cada resultado del dado su propio valor numérico. Entonces, $X(\omega) = \omega$ para cada $\omega \in \Omega$.
### ¿Para qué se usa?
Las variables aleatorias permiten modelar fenómenos inciertos, facilitando la toma de decisiones en estadística e ingeniería. Son fundamentales en la inferencia estadística, la teoría de colas, la teoría de juegos, finanzas, inteligencia artificial, entre otros campos.

# Sección 2: Función de Distribución
# =================================================

## Definición y conceptos básicos

La función de distribución acumulada (FDA) de una variable aleatoria $X$, denotada por $F(x)$, se define como:
$$
F(x) = P(X \leq x)
$$
Representa la probabilidad acumulada de que la variable aleatoria tome un valor menor o igual a $x$. La FDA satisface las siguientes propiedades:

1. **Monotonía**: $F(x)$ es una función no decreciente.
2. **Límites**:
$$
\lim_{x \to -\infty} F(x) = 0 \quad \text{y} \quad \lim_{x \to \infty} F(x) = 1.
$$
3. **Continuidad por la derecha**: $F(x)$ es continua por la derecha, es decir, $F(x) = F(x^+)$.

Para una variable aleatoria discreta, $F(x)$ es una función escalonada, mientras que para una variable aleatoria continua es una función suave que se relaciona con la función de densidad de probabilidad mediante derivadas.




````python
# Ejemplo práctico: Distribución para el lanzamiento de un dado
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 7)
p = np.ones_like(x) / len(x)  # Probabilidades de cada cara del dado
F = np.cumsum(p)  # Función de distribución acumulada

# Mostrar probabilidades y la función de distribución acumulada
print("Probabilidades:\n", p)
print("Función de distribución acumulada:\n", F)

# Gráfica de la función de distribución
plt.step(x, F, where="post", label="Función de Distribución")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.title("Función de distribución acumulada para un dado")
plt.legend()
plt.grid()
plt.show()


````

# Sección 3: Función de Densidad de Probabilidad
# =================================================
## Definición y ejemplos
# Función de Densidad de Probabilidad (FDP)
La función de densidad de probabilidad (FDP) describe cómo se distribuyen las probabilidades en el rango de valores de una **variable aleatoria continua**. Formalmente, para una variable aleatoria continua $X$ con función de distribución acumulada $F(x)$, la FDP $f(x)$ se define como:

$$
f(x) = \frac{dF(x)}{dx}, \quad \text{para valores de } x \text{ donde } F(x) \text{ es diferenciable.}
$$

La FDP tiene las siguientes propiedades:

1. $ f(x) \geq 0$ para todo \( x \).
2. La integral de \( f(x) \) sobre todo el dominio de \( X \) es igual a 1:
$$
\int_{-\infty}^{\infty} f(x) \, dx = 1.
$$
3. La probabilidad de que \( X \) tome un valor en un intervalo \([a, b]\) se calcula como:
$$
P(a \leq X \leq b) = \int_a^b f(x) \, dx.
$$


````python
# Ejemplo práctico: Distribución normal
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución normal
mu, sigma = 0, 1  # Media y desviación estándar

# Generación de valores para x
x = np.linspace(-4, 4, 1000)

# Cálculo de la función de densidad de probabilidad (FDP)
fx = (1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * ((x - mu) / sigma)**2)

# Graficar la FDP
plt.plot(x, fx, label="Distribución Normal")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Función de densidad de probabilidad (Distribución Normal)")
plt.legend()
plt.grid()
plt.show()
````

# Sección 4: Función de Masa de Probabilidad
# =================================================
## Definición y conceptos básicos
# Función de Masa de Probabilidad (FMP)

La función de masa de probabilidad (FMP) es una función que describe la distribución de probabilidad de una **variable aleatoria discreta**. Para una variable aleatoria discreta $X$, la FMP se denota como $p(x)$ y está definida como:

$$
p(x) = P(X = x),
$$

donde \( x \) es un valor que la variable aleatoria puede tomar.

La FMP cumple con las siguientes propiedades fundamentales:

1. $ p(x) \geq 0 $ para todo \( x \).
2. La suma de \( p(x) \) sobre todos los posibles valores de \( X \) es igual a 1:
$$
\sum p(x) = 1.
$$


````python
# Ejemplo: Función de Masa de Probabilidad para un dado justo
import numpy as np
import matplotlib.pyplot as plt

# Descripción del ejemplo
print("Ejemplo: Considere un dado justo. La variable aleatoria X representa el resultado de un lanzamiento.")
print("Los valores posibles de X son 1, 2, 3, 4, 5 y 6, cada uno con una probabilidad igual de 1/6.")
print("La FMP está dada por:")
print("p(x) = 1/6 para x = 1, 2, 3, 4, 5, 6.\n")

# Valores posibles y sus probabilidades
valores_dado = np.array([1, 2, 3, 4, 5, 6])
probabilidades_dado = np.ones_like(valores_dado) / len(valores_dado)

# Gráfico de la FMP
plt.stem(valores_dado, probabilidades_dado, basefmt=" ")
plt.title("Función de Masa de Probabilidad para un Dado")
plt.xlabel("Valores del Dado")
plt.ylabel("p(x)")
plt.grid()
plt.show()
````

# Sección 5: Valor Esperado, Varianza y Momentos
# =================================================
# Valor Esperado y Varianza

El **valor esperado** de una variable aleatoria $X$, denotado como $E[X]$, representa el promedio ponderado de todos los valores posibles de $X$, donde los pesos son las probabilidades asociadas. Matemáticamente, está definido como:

- Para variables aleatorias discretas:
  $$
  E[X] = \sum x \cdot p(x).
  $$

- Para una variable continua:
  $$
  E[X] = \int x \cdot f(x) \, dx,
  $$
  donde $f(x)$ es la función de densidad de probabilidad.

La **varianza**, denotada como $\text{Var(X)}$, mide la dispersión de la variable alrededor de su valor esperado. Se define como:

- Para variables discretas:
  $$
  \text{Var}(X) = \sum \left( x - E[X] \right)^2 \cdot p(x).
  $$

- Para variables continuas:
  $$
  \text{Var}(X) = \int (x - E[X])^2 \cdot f(x) \, dx,
  $$
  donde $f(x)$ es la función de densidad de probabilidad de la variable continua.






````python
## Cálculo del valor esperado y la varianza para un dado justo

import numpy as np

# Valores y probabilidades del dado
valores_dado = np.array([1, 2, 3, 4, 5, 6])
probabilidades_dado = np.ones_like(valores_dado) / len(valores_dado)

# Cálculo del valor esperado
esperanza_dado = np.sum(valores_dado * probabilidades_dado)

# Cálculo de la varianza
varianza_dado = np.sum(((valores_dado - esperanza_dado)**2) * probabilidades_dado)

# Resultados
print(f"Valor esperado del dado: {esperanza_dado:.2f}")
print(f"Varianza del dado: {varianza_dado:.2f}")
````

````python
import numpy as np
from scipy.stats import norm

# Parámetros de la distribución normal estándar
mu, sigma = 0, 1

# Función de densidad de probabilidad para una distribución normal
def f(x):
    return (1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * ((x - mu) / sigma)**2)

# Rango de valores para x
x = np.linspace(-4, 4, 1000)

# Cálculo del valor esperado para la distribución normal
esperanza_continua = np.sum(x * f(x)) * (x[1] - x[0])  # Aproximación por suma

# Cálculo de la varianza para la distribución normal
varianza_continua = np.sum(((x - esperanza_continua)**2) * f(x)) * (x[1] - x[0])  # Aproximación por suma

# Resultados para la variable continua
print(f"Valor esperado para la distribución normal: {esperanza_continua:.2f}")
print(f"Varianza para la distribución normal: {varianza_continua:.2f}")
````

# Sección 6: Definición y Propiedades del Valor Esperado
# =================================================
## Conceptos y propiedades

# Propiedades del Valor Esperado

El **valor esperado** tiene varias propiedades importantes que facilitan su cálculo en combinaciones de variables aleatorias. Entre las más relevantes están:

1. **Linealidad**:
   $$ E[aX + b] = aE[X] + b, $$
   donde \( a \) y \( b \) son constantes. Esto significa que si multiplicamos una variable aleatoria por una constante \( a \) y le sumamos una constante \( b \), el valor esperado se comportará de manera lineal.

2. **Suma de Variables Independientes**:
   $$ E[X + Y] = E[X] + E[Y], $$
   No importa $X$ y $Y$ son variables aleatorias **dependientes**.

3. **Multiplicación por Constantes**:
   $$ E[cX] = cE[X], $$
   donde \( c \) es una constante. Si multiplicamos una variable aleatoria \( X \) por una constante \( c \), el valor esperado se multiplicará por la misma constante \( c \).

4. **Segundo momento con respecto a cero**:
   $$ E[X^2] = \text{Var}(X) + (E[X])^2, $$


5. **Esperanza de la Función de una Variable Aleatoria**:
   Si $g(X)$ es una función de la variable aleatoria $X$, entonces el valor esperado de $g(X)$ es:
   $$ E[g(X)] = \int g(x) f(x) \, dx $$
   $$ E[g(X)] = \sum_{i} g(x_i) p(x_i)$$

## Sección 7: Cálculo de la Varianza

### Definición y ejemplos

La **varianza** mide la dispersión de una variable aleatoria respecto a su valor esperado. Se define como:
$$
\text{Var}(X) = E[(X - E[X])^2],
$$
donde \( E[X] \) es el valor esperado de \( X \).

### Propiedades de la Varianza

Las propiedades de la varianza son fundamentales para simplificar el cálculo y análisis de combinaciones de variables aleatorias. Entre las más relevantes se encuentran:

1. **No negatividad**:
   $$
   \text{Var}(X) \geq 0.
   $$
   La varianza nunca es negativa, ya que es una medida de dispersión cuadrática.

2. **Varianza de una constante**:
   $$
   \text{Var}(c) = 0,
   $$
   donde \( c \) es una constante. Esto tiene sentido porque una constante no tiene dispersión, siempre toma el mismo valor.

3. **Linealidad de la Varianza**:
   Si \( a \) es una constante:
   $$
   \text{Var}(aX) = a^2 \cdot \text{Var}(X).
   $$

4. **Suma de Variables Independientes**:
   Si \( X \) y \( Y \) son independientes:
   $$
   \text{Var}(X + Y) = \text{Var}(X) + \text{Var}(Y).
   $$

5. **Varianza de una Suma de Variables Correlacionadas**:
   Si \( X \) y \( Y \) no son independientes, entonces:
   $$
   \text{Var}(X + Y) = \text{Var}(X) + \text{Var}(Y) + 2 \cdot \text{Cov}(X, Y),
   $$
   donde $\text{Cov}(X, Y)$ es la **covarianza** entre \( X \) y \( Y \).

6. **Varianza de la Diferencia de Variables**:
   Si \( X \) y \( Y \) son variables aleatorias:
   $$
   \text{Var}(X - Y) = \text{Var}(X) + \text{Var}(Y) - 2 \cdot \text{Cov}(X, Y).
   $$

7. **Varianza de una combinación lineal de variables**:
   Si \( a \) y \( b \) son constantes y \( X \) y \( Y \) son variables aleatorias:
   $$
   \text{Var}(aX + bY) = a^2 \cdot \text{Var}(X) + b^2 \cdot \text{Var}(Y) + 2ab \cdot \text{Cov}(X, Y).
   $$

---

# Sección 8: Momentos de una Variable Aleatoria
==========================================================
## Definición y cálculo


# Momentos de una Variable Aleatoria

Los **momentos** de una variable aleatoria son medidas que resumen características de su distribución. El momento de orden $r$ conm respecto al centro se define como $E[X^r]$ para $ r = 1, 2, \dots $.

## Cálculo de Momentos para un Dado

El código siguiente calcula los momentos de orden 1 a 4 para un dado justo.

```python
import numpy as np

# Valores y probabilidades del dado
valores_dado = np.array([1, 2, 3, 4, 5, 6])
probabilidades_dado = np.ones_like(valores_dado) / len(valores_dado)

# Cálculo de los momentos
momentos_dado = [np.sum((valores_dado**r) * probabilidades_dado) for r in range(1, 5)]

# Mostrar los resultados
for r, momento in enumerate(momentos_dado, start=1):
    print(f"Momento de orden {r}: {momento:.2f}")
```

### Explicación:

En este código, se calcula el valor esperado de \( X^r \) para \( r = 1, 2, 3, 4 \), utilizando la fórmula general de los momentos.

- El **primer momento** (de orden 1) es el valor esperado \( E[X] \), que da el promedio de los valores posibles del dado.
- El **segundo momento** (de orden 2) está relacionado con el valor esperado del cuadrado de la variable $E[X^2]$, el cual se utiliza para calcular la **varianza**.
- Los **momentos de orden superior** se utilizan para describir otras características de la distribución, como la asimetría o la curtosis.

---

## Función Generadora de Momentos (FGM)

La **función generadora de momentos (FGM)** es una herramienta matemática utilizada para caracterizar completamente la distribución de una variable aleatoria. Está definida como:
$$
M(t) = E[e^{tX}],
$$
donde $ t \in \mathbb{R} $ y \( E \) es el valor esperado.

### Importancia de la FGM:

La importancia de la FGM radica en que, al derivarla con respecto a $t$ y evaluarla en $t = 0$, se obtienen los momentos de la variable aleatoria. Es decir:

1. $M'(0) = E[X]$, el primer momento (valor esperado).
2. $M''(0) = E[X^2]$, el segundo momento, a partir del cual se puede calcular la varianza.
3. De manera general, el momento de orden $r$ está dado por $M^{(r)}(0) = E[X^r]$.

La FGM existe si el valor esperado de $e^{tX}$ converge para un intervalo de valores de $t$ alrededor de 0.

---

## Propiedades de la FGM

1. **Propiedad 1:**
  $$
   M(0) = 1, \quad \text{ya que} \ e^{0 \cdot X} = 1 \ \text{para cualquier} \ X.
  $$

2. **Propiedad 2:**
   La FGM de una suma de variables aleatorias independientes es el producto de sus FGMs individuales:
   $$
   M_{X+Y}(t) = M_X(t) \cdot M_Y(t).
   $$

3. **Propiedad 3:**
   La derivada de orden $r$ $M(t)$ evaluada en  $t = 0$ proporciona el momento de orden $r$ de la variable aleatoria:
   $$
   M^{(r)}(0) = E[X^r].
   $$

4. **Propiedad 4:**
   La FGM de una constante es:
   $$
   M_c(t) = e^{ct}.
   $$

5. **Propiedad 5:**
   Si $X$ y $Y$ son independientes, entonces la FGM de su suma es el producto de sus FGMs:
   $$
   M_{X+Y}(t) = M_X(t) \cdot M_Y(t).
   $$

6. **Propiedad 6:**
   La FGM de una constante multiplicada por una variable aleatoria es:
   $$
   M_{aX}(t) = M_X(at).
   $$
   Donde $a$ es una constante.

````python
# Ejemplo práctico: FGM para un dado justo
print("Ejemplo: Considere un dado justo con valores X = {1, 2, 3, 4, 5, 6}. La probabilidad asociada a cada valor es 1/6. Queremos calcular la FGM de esta variable aleatoria.\n")

# Definición de los valores del dado y sus probabilidades
valores_dado = np.array([1, 2, 3, 4, 5, 6])
probabilidades_dado = np.ones_like(valores_dado) / len(valores_dado)

# Parámetro t
print("Calculemos la FGM para t = 0.5.\n")
t = 0.5
fgm_dado = np.sum(np.exp(t * valores_dado) * probabilidades_dado)
print(f"M(0.5) = {fgm_dado:.4f}\n")

# Cálculo de la FGM para diferentes valores de t
t_values = np.linspace(-0.5, 1, 100)
fgm_values = [np.sum(np.exp(t * valores_dado) * probabilidades_dado) for t in t_values]

# Gráfico de la FGM
plt.plot(t_values, fgm_values, label="FGM del dado")
plt.xlabel("t")
plt.ylabel("M(t)")
plt.title("Función Generadora de Momentos (FGM) para un Dado Justo")
plt.legend()
plt.grid()
plt.show()
````

# Sección 9: Distribuciones discretas
# =================================================
Se presentan algunas de las distribuciones discretas más usadas

## 1. Distribución uniforme

Se dice que una variable aleatoria $X$ tiene distribución discreta uniforme de parámetro $N$, donde $N$ es un entero positivo, si su función de masa de probabilidad está dada por
$$
P(X = x) =
\begin{cases}
\frac{1}{n} & \text{si } x = 1,2, ..., N \\
0 & \text{en otro caso}
\end{cases}
$$

### Propiedades
1.   $ E[X] = \frac{N+1}{2} $
2.   $ var[x] = \frac{N^2-1}{12}$
3. $ M_X(t) = \sum_{k=1}^N \frac{1}{N} e^{tk}$

### Ejemplo
Suponga que se lanza un dado justo de 6 caras, donde $X$ es el número que aparece en el lanzamiento. En este caso se tiene la función de densidad
$$
P(X = x) =
\begin{cases}
\frac{1}{6} & \text{si } x = 1,2, ..., 6 \\
0 & \text{en otro caso}
\end{cases}
$$

````python
import matplotlib.pyplot as plt

# Parámetros de la distribución uniforme discreta
a, b = 1, 6 # Rango de valores
n = b - a + 1  # Número de posibles valores

# Valores de x y probabilidades
x = range(a, b + 1)
p = [1 / n] * n  # Probabilidad constante

# Graficar
plt.stem(x, p, basefmt=" ")
plt.title("Función de masa de probabilidad")
plt.xlabel("x (valores posibles)")
plt.ylabel("P(X = x)")
plt.xticks(x)
plt.ylim(0, 1)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
````

## 2. Distribución Bernoulli
Un experimento de Bernoulli es aquel que tiene exactamente dos resultados posibles, "Éxito" (1) y "Fracaso" (0). Se dice que una variable aleatoria $X$ tiene distribución bernoulli de parámetro $p$, $X \sim \text{Ber}(p)$, si su función de masa de probabilidad está dada por

$$P(X = x) =
\begin{cases}
p & \text{si } x = 1 \\
1 - p & \text{si } x = 0
\end{cases}
$$

donde $𝑝$ es la probabilidad de éxito y $1−p$ es la probabilidad de fracaso. Dado que $p$ es una probabilidad, el soporte de la variable será $(0,1)$ y su función de distribución será dada por

$$F(x) =
\begin{cases}
0 & \text{si } x < 0 \\
1 - p & \text{si } 0 \leq x < 1 \\
1 & \text{si } x \geq 1
\end{cases}
$$

### Propiedades
1.   $ E[X] = p $
2.   $ var[x] = p(1-p)$
3. $ M_X(t) = (1-p)+pe^t$

### Ejemplo
Supongamos que en una empresa repartidora asignamos \( X = 1 \) si una entrega es realizada a tiempo (éxito) y \( X = 0 \) si se entrega después (fracaso). La probabilidad de que un paquete sea entregado dentro de los límites de tiempo establecidos es del 80%. Entonces, la variable aleatoria \( X \) sigue una distribución de Bernoulli con \( p = 0.8\).



````python
# Parámetro de la distribución de Bernoulli
p = 0.8 # Probabilidad de que salga 1 (éxito)

# Valores posibles de la variable aleatoria X (0 y 1)
x = [0, 1]
# Probabilidades asociadas
probabilidades = [1 - p, p]

# Graficar la función de densidad
plt.stem(x, probabilidades, basefmt=" ")
plt.title("Función de probabilidad distribución Bernoulli")
plt.xlabel("X (0 = Fracaso, 1 = Éxito)")
plt.ylabel("P(X = x)")
plt.xticks(x)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
````

## 3. Distribución Binomial
Es utilizada cuando se realizan experimentos repetidos y los resultados son binarios (éxito o fracaso). Es una extensión de la distribución de Bernoulli cuando se realizan múltiples ensayos independientes. La variable $X$ cuenta el número de éxitos obtenidos en total y se denota $X \sim Bin(n,p)$

Se dice que una variable aleatoria $X$ tiene distribución binomial de parámetros $n$ y $p$, si su función de masa de probabilidad está dada por

$$P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}$$

donde $p$ es la probabilidad de éxito en un solo ensayo

### Propiedades
1.   $ E[X] = np $
2.   $ var[x] = np(1-p)$
3. $ M_X(t) = ((1-p)+pe^t)^n$

### Ejemplo
Supongamos que lanzamos una moneda 10 veces, y queremos saber la probabilidad de obtener exactamente 6 caras (éxitos). Si la probabilidad de que salga cara en cada lanzamiento es $𝑝=0.5$, podemos calcular la probabilidad usando la fórmula de la distribución binomial.
$$
P(X = 6) = \binom{10}{6} (0.5)^6 (0.5)^{10-6} = \binom{10}{6} (0.5)^{10} \approx 0.205
$$


````python
from scipy.stats import binom

# Parámetros de la distribución binomial
n = 10  # Número de lanzamientos
p = 0.5  # Probabilidad de éxito (cara)

k = np.arange(0, n+1)

# Probabilidad de obtener k éxitos
probabilidades = binom.pmf(k, n, p)

# Graficar la función de probabilidad con líneas verticales
plt.vlines(k, 0, probabilidades, lw=2)
plt.title('Distribución Binomial de 10 lanzamientos de una moneda')
plt.xlabel('Número de éxitos (caras)')
plt.ylabel('Probabilidad')
plt.xticks(k)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
````

## 4. Distribución poisson
Describe el número de eventos que ocurren en un intervalo de tiempo o espacio dado, bajo las siguientes condiciones:

*   Los eventos ocurren de manera independiente entre sí.
*   Los eventos ocurren a una tasa promedio constante $\lambda$ (lambda), que es el número esperado de eventos en el intervalo.

Se dice que una variable aleatoria $X$ tiene distribución Poisson de parámetro  $\lambda > 0$, si su función de densidad está dada por:

$$P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}$$

donde $\lambda$ es la tasa promedio de ocurrencia de eventos y $k$ es el número de eventos. Se denota como $X \sim P(\lambda)$

### Propiedades
1.   $ E[X] = \lambda $
2.   $ var[x] = \lambda$
3. $ M_X(t) =  \text{exp}(\lambda(e^t-1))$

### Ejemplo
Si el número promedio de autos que pasan por una intersección en una hora es $ \lambda = 5 $, ¿cuál es la probabilidad de que no pase ningún auto $( k = 0 )$?

$$
P(X = 0) = \frac{5^0 e^{-5}}{0!} = \frac{1 \cdot e^{-5}}{1} = e^{-5} \approx 0.0067
$$

````python
from scipy.stats import poisson

# Parámetros de la distribución de Poisson
lambda_ = 5  # Tasa promedio de ocurrencia
k = np.arange(0, 15)  # Valores posibles de número de eventos

# Probabilidad de obtener k eventos
probabilidades = poisson.pmf(k, lambda_)

# Graficar la función de masa de probabilidad
plt.stem(k, probabilidades, basefmt=" ")
plt.title('Funcion de probabilidad Poisson (λ = 5)')
plt.xlabel('Número de autos (k)')
plt.ylabel('Probabilidad')
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(k)
plt.show()
````

Al analizar la gráfica de la función de densidad de una variable aleatoria
con distribución Poisson, se observa que las probabilidades individuales son
cada vez más pequeñas a medida que la variable toma valores cada vez más
grandes. Ésta es, precisamente, una característica general de la distribución
Poisson.

Así mismo, la distribución de Poisson es una forma límite de la distribución binomial, cuando $n$ es suficientemente grande y $p$ suficiente y adecuadamente pequeño. Empíricamente se ha establecido, que la aproximación se puede
aplicar con seguridad si $n > 100, p< 0.01, np < 20$.

**Teorema**
Si $p(n)$ es una sucesión con $0< p(n)< 1$ y $n(p(n)) \longrightarrow \lambda$ cuando $n \longrightarrow \infty$, entonces

$$
Ber_{n, p(n)}(k):=\binom{n}{k}(p(n))^k(1-p(n))^{n-k} \underset{n \rightarrow \infty}{\longrightarrow} e^{-\lambda} \frac{\lambda^k}{k!}=: P(\lambda)(k)
$$

### Ejemplo
 En un auditorio se encuentran 135 estudiantes. La probabilidad de que uno de los estudiantes se encuentre hoy de cumpleaños es igual $a \frac{1}{365} \cdot$ ¿Cuál es la probabilidad de que en el auditorio dos o más estudiantes estén hoy de cumpleaños?

 Sea $X:=$ "número de estudiantes que están de cumpleaños hoy". Se sabe que $X \sim Ber\left(135, \frac{1}{365}\right)$. Ésta última distribución se puede aproximar por una Poisson con parámetro $\lambda=\frac{27}{73}$. Por lo tanto,

$$
\begin{aligned}
P(X & \geq 2)=1-P(X=0)-P(X=1) \\
& =1-e^{-\frac{27}{73}}-\frac{27}{73} e^{-\frac{27}{73}} \\
& =5.3659 \times 10^{-2} .
\end{aligned}
$$

## 5. Distribución Hipergeométrica

Describe la probabilidad de obtener un número específico de éxitos en una muestra de tamaño $n$, extraída sin reemplazo de una población finita de tamaño $N$.

Se dice que una variable aleatoria $X$ tiene distribución hipergeométrica de parámetros $n$, $R$ y $N$, si su función de masa de probabilidad está dada por:

$$
P(X = x) = \frac{\binom{R}{x} \binom{N-R}{n-x}}{\binom{N}{n}}
$$

donde:
- \( N \) es el tamaño total de la población.
- \( R \) es el número total de éxitos en la población.
- \( n \) es el tamaño de la muestra.

siendo $N$ un entero positivo, $R$ un entero no negativo menor o igual a
$N$ y $n$ un entero positivo menor o igual a $N$.

### Propiedades
1.   $  E[X] = \frac{nR}{N} $
2.   $ Var[x]  = \frac{nR(N-R)(N-n)}{N^2(N-1)}$
3. $ M_X(t) =  \text{exp}(\lambda(e^t-1))$

### Ejemplo

Supongamos que tenemos una urna con 20 bolas: 8 rojas y 12 azules. Queremos extraer una muestra de 5 bolas sin reemplazo. ¿Cuál es la probabilidad de que obtengamos exactamente 3 bolas rojas?

En este caso:
- $ N = 20$ (total de bolas),
- $K = 8$ (bolas rojas),
- $n = 5$ (tamaño de la muestra),
- $k = 3 $ (número de bolas rojas en la muestra).

Entonces, la probabilidad de obtener exactamente 3 bolas rojas es:

$$
P(X = 3) = \frac{\binom{8}{3} \binom{12}{2}}{\binom{20}{5}} \approx 0.2381
$$



````python
from scipy.stats import hypergeom

# Parámetros
N = 20  # Tamaño total de la población
R = 8   # Número de éxitos en la población
n = 5   # Tamaño de la muestra

# Valores de k (número de éxitos en la muestra)
k = np.arange(0, n+1)

# Calcular la PMF de la distribución hipergeométrica
pmf = hypergeom.pmf(k, N, R, n)

# Graficar
plt.stem(k, pmf, basefmt=" ")
plt.title("Función de probabilidad Distribución Hipergeométrica")
plt.xlabel("Número de éxitos en la muestra (k)")
plt.ylabel("P(X = k)")
plt.xticks(k)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
````

Si el tamaño de la población $N$ es grande, en comparación con el tamaño de la muestra $n$, entonces, la distribución hipergeométrica puede aproximarse por una distribución binomial. Más precisamente se tiene el siguiente resultado:

**Teorema**
Sea $0< p< 1$. Si $N, R \longrightarrow \infty$ de tal forma que $\frac{R}{N} \longrightarrow p$, entonces:
$$
Hg (n, R, N)(k):=\frac{\binom{R}{k}\binom{N-R}{n-k}}{\binom{N}{n}} \longrightarrow Ber(n, p)(k):=\binom{n}{k} p^k(1-p)^{n-k}
$$

### Ejemplo
 En una ciudad de 2 millones de habitantes, se tiene que el $60\%$ pertenecen al partido político $A$. Se eligen al azar $100$ personas. La distribución del número de personas, entre los $100$ elegidos, que pertenecen al partido $A$ es una distribución hipergeométrica de parámetros $100$, $1.200.000$ y $2.000.000$ . Aplicando el resultado anterior se puede aproximar esta distribución por una binomial de parámetros $n=1000, p=0.6$. Así, por ejemplo, la probabilidad de que entre los $100$ elegidos exactamente $40$ pertenezcan al partido $A$ es igual a:

$$
\binom{100}{40}(0.6)^{40}(0.4)^{60}=2.4425 \times 10^{-5} .
$$

## 1. Distribución Uniforme continua
La \textbf{distribución uniforme continua} es una distribución de probabilidad en la que todos los valores dentro de un intervalo específico son igualmente probables. Si una variable aleatoria $X$ sigue una distribución uniforme continua en el intervalo $[a, b]$, se dice que $X \sim U(a, b)$.

Se dice que una variable aleatoria X está distribuida uniformemente sobre el intervalo $[a, b]$, con $a < b$ números reales, si su función de densidad está dada por

$$
f_X(x) =
\begin{cases}
 \frac{1}{b - a} & \text{si } a \leq x \leq b\\
 0 & \text{si } x < a , x > b
 \end{cases}
$$

La función de distribución acumulada (CDF) de una distribución uniforme continua es:

$$
F_X(x) = \begin{cases}
0, & \text{si } x < a \\
\frac{x - a}{b - a}, & \text{si } a \leq x \leq b \\
1, & \text{si } x > b
\end{cases}
$$

### Propiedades
1.   $  E[X] =  \frac{a + b}{2} $
2.   $ Var[x]  = \frac{(b - a)^2}{12}$

Supongamos que tenemos una distribución uniforme continua en el intervalo \([a, b] = [1, 5]\). La función de densidad de probabilidad será:

$$
f_X(x) = \frac{1}{5 - 1} = \frac{1}{4}, \quad \text{para } 1 \leq x \leq 5
$$


````python
from scipy.stats import uniform

# Parámetros
a = 1  # Inicio del intervalo
b = 5  # Fin del intervalo

# Valores para graficar
x = np.linspace(a - 1, b + 1, 1000)  # Generamos puntos alrededor de [a, b]
pdf = uniform.pdf(x, loc=a, scale=b - a)  # PDF de la distribución uniforme continua
cdf = uniform.cdf(x, loc=a, scale=b - a)  # CDF de la distribución uniforme continua

# Graficar PDF
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF', color='blue')
plt.fill_between(x, pdf, alpha=0.2, color='blue')
plt.title("Función de Densidad")
plt.xlabel("x")
plt.ylabel("f_X(x)")
plt.grid(True)

# Graficar CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF', color='red')
plt.title("Función de Distribución Acumulada")
plt.xlabel("x")
plt.ylabel("F_X(x)")
plt.grid(True)

plt.tight_layout()
plt.show()
````

### Ejemplo
Se ha determinado que el tiempo que tarda una maquina en hacer una determinada tarea sigue una distribución uniforme sobre el intervalo de 30 segundos a 40 segundos. Determine la proporción de veces que la tarea tomará más de
36 segundos para ser completada.

La función de distribución acumulada (CDF) es:

$$
F_X(x) = \frac{x - 30}{40 - 30} = \frac{x - 30}{10}
$$

Evaluando en \( x = 36 \):

$$
F_X(36) = \frac{36 - 30}{40 - 30} = \frac{6}{10} = 0.6
$$

Por lo tanto:

$$
P(X > 36) = 1 - F_X(36) = 0.4
$$


## 2. Distribución Exponencial

Se utiliza para modelar el tiempo entre eventos en un proceso de Poisson, es decir, cuando los eventos ocurren de manera aleatoria y a una tasa constante. La variable X toma un valor positivo. Generalmente se usa para modelar
tiempos de espera o de vida (duraciones). Se denota $X \sim Exp(\lambda)$ donde el parámetro $\lambda$ es el número promedio de sucesos por unidad de tiempo.

Se dice que la variable aleatoria $X$ tiene distribución exponencial si su función de densidad está dada por:
$$
f_X(x)=\left\{\begin{array}{l}
\lambda e^{-\lambda x}, \quad x \geq 0 \\
0, \text { en otro caso }
\end{array}\right.
$$

La función de distribución estará dada por:
 $$ F_X(x)=\left\{\begin{array}{l}0, \quad x< 0 \\ 1- e^{-\lambda x}, \quad x \geq 0\end{array}\right.$$

### Propiedades
1. $E[X]=\frac{1}{\lambda}$.
2. $Var(X)=\frac{1}{\lambda^2}$.
3. $m_X(t)=\frac{\lambda}{\lambda-t}, \quad t < \lambda$.

Un aspecto importante de la distribución exponencial es que tiene la propiedad de "pérdida de memoria". Esto significa que la probabilidad de que el tiempo de espera hasta el próximo evento sea mayor que $t+s$, dado que ya ha pasado un tiempo $t$, es la misma que la probabilidad de que el tiempo de espera hasta el primer evento sea mayor que $s$.

Sea $X$ una variable aleatoria tal que $X\sim\operatorname{Exp}(\lambda)$ entonces para cualesquiera $x,y\geq 0$

$$\operatorname{P}[X>x+y|X>y]=\operatorname{P}[X>x]$$

Esto puede demostrarse fácilmente pues

$$\begin{align}
    \operatorname{P}[X>x+y|X>y]
    &=\frac{\operatorname{P}[X>x+y\cap X>y]}{\operatorname{P}[X>y]} \\
    &=\frac{\operatorname{P}[X>x+y]}{\operatorname{P}[X>y]} \\
    &=\frac{e^{-\lambda(x+y)}}{e^{-\lambda y}} \\
    &=e^{-\lambda x} \\
    &=\operatorname{P}[X>x]
\end{align}$$

### Ejemplo
Supongamos que en una central de atención telefónica, el tiempo entre llegadas de llamadas sigue una distribución exponencial con una tasa de $ \lambda = 0.2 $ llamadas por minuto. Esto significa que, en promedio, llega una llamada cada 5 minutos ($ \mu = \frac{1}{\lambda} = 5 $ minutos). ¿Cuál es la probabilidad de que pase menos de 3 minutos entre dos llamadas consecutivas?


$$
F_X(x) = 1 - e^{-\lambda x}
$$

$$
P(X < 3) = 1 - e^{-0.2 \times 3} = 1 - e^{-0.6}
$$

Calculando este valor:

$$
P(X < 3) = 1 - e^{-0.6} \approx 1 - 0.5488 = 0.4512
$$

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parámetro de la distribución (lambda)
lambda_exp = 2  # Tasa de ocurrencia de eventos

# Rango de valores para t
t = np.linspace(0, 3, 1000)  # Valores desde 0 hasta 3 con 1000 puntos

# Función de densidad (PDF): f(t) = λ * e^(-λt)
pdf = lambda_exp * np.exp(-lambda_exp * t)

# Función de distribución acumulada (CDF): P(T ≤ t) = 1 - e^(-λt)
cdf = 1 - np.exp(-lambda_exp * t)

# Crear la figura y las subgráficas
plt.figure(figsize=(10, 5))

# Gráfico de la densidad (PDF)
plt.subplot(1, 2, 1)
plt.plot(t, pdf, label="Densidad (PDF)", color="blue")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.title("Función de Densidad Exponencial (PDF)")
plt.legend()
plt.grid()

# Gráfico de la distribución acumulada (CDF)
plt.subplot(1, 2, 2)
plt.plot(t, cdf, label="Distribución Acumulada (CDF)", color="red")
plt.xlabel("t")
plt.ylabel("P(T ≤ t)")
plt.title("Función de Distribución Acumulada (CDF)")
plt.legend()
plt.grid()

# Mostrar la figura
plt.tight_layout()
plt.show()
````

# Una variable con distribución exponencial no tiene memoria

La **propiedad de sin memoria** es una característica clave de la distribución exponencial y, por extensión, del **Proceso de Poisson**. Esta propiedad establece que el tiempo adicional que hay que esperar para el próximo evento **no depende del tiempo ya transcurrido sin eventos**.

---

## Definición Formal

Si $T$ es una variable aleatoria que representa el **tiempo hasta el próximo evento** y sigue una distribución exponencial con parámetro $\lambda$, entonces cumple la propiedad de sin memoria:

$$
P(T > t + s \mid T > t) = P(T > s)
$$

para todo $t, s \geq 0$.

Esto significa que, dado que el evento aún no ha ocurrido hasta el tiempo $t$, la probabilidad de que tarde al menos $s$ unidades más es la misma que si se comenzara a contar desde cero.

---

## Explicación Intuitiva

- Si un proceso sigue la propiedad de sin memoria, **el tiempo ya esperado no proporciona información sobre cuándo ocurrirá el siguiente evento**.
- Esto es característico de procesos donde los eventos ocurren de forma completamente aleatoria y sin influencia del pasado.
- Por ejemplo, si el tiempo de espera entre llamadas en un centro de atención sigue una distribución exponencial, **el tiempo que has estado esperando no cambia la expectativa de cuánto más tendrás que esperar**.

---

## Demostración de la Propiedad

Partimos de la función de supervivencia de la distribución exponencial:

$$
P(T > t) = e^{-\lambda t}.
$$

Ahora, aplicamos la definición de probabilidad condicional:

$$
P(T > t + s \mid T > t) = \frac{P(T > t + s)}{P(T > t)}.
$$

Sustituyendo los valores:

$$
\frac{e^{-\lambda (t+s)}}{e^{-\lambda t}} = e^{-\lambda s}.
$$

Observamos que el resultado **no depende de $t$**, solo de $s$, lo que demuestra la propiedad de sin memoria.

---

## Relación con el Proceso de Poisson

La propiedad de sin memoria es una consecuencia directa del **Proceso de Poisson**:

- En un proceso de Poisson con tasa $\lambda$, el **tiempo entre eventos** sigue una distribución exponencial Exp($\lambda$).
- Como los tiempos entre eventos son **independientes**, el tiempo de espera restante **no depende del tiempo ya transcurrido sin eventos**.

Esto hace que el proceso de Poisson sea un modelo ideal para eventos que ocurren de forma completamente aleatoria, como:

- Llamadas telefónicas en una central.
- Llegadas de clientes a un banco.
- Errores en un sistema informático.

---

## Ejemplo Numérico

Si en un sistema de atención al cliente llegan llamadas a una tasa de $\lambda = 5$ llamadas por minuto:

- **¿Cuál es la probabilidad de que una llamada tarde más de 2 minutos en llegar, dado que ya han pasado 3 minutos sin recibir ninguna?**

Usamos la propiedad de sin memoria:

$$
P(T > 2 \mid T > 3) = P(T > 2) = e^{-5(2)} = e^{-10} \approx 4.54 \times 10^{-5}.
$$

El tiempo ya transcurrido no influye en la probabilidad.

## 2. Distribución Normal

La distribución normal es una de las más importantes y de mayor uso
tanto en la teoría de la probabilidad, como en la teoría estadística.


Se dice que la variable aleatoria $X$ tiene distribución normal de parámetros $\mu y \sigma$, donde $\mu$ es un número real $y \sigma$ es un número real positivo, si su función de densidad está dada por:

$$
f(x)=\frac{1}{\sigma \sqrt{2 \pi}} \exp \left[-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2\right] ; x \in \mathbb{R}
$$

donde $μ$ es la media de la distribución (parámetro de localización) y $𝜎$ es la desviación estándar (parámetro de escala)

### Propiedades
1. $E[X]=\mu$.
2. $Var(X)=\sigma^2$.
3. $m_X(t)=\exp \left[\mu t+\frac{\sigma^2 t^2}{2}\right]$.

La función de distribución no tiene una forma análitica, sus resultados pueden obtenerse computacionalmente. Suponga el caso en que $\mu =50, \sigma = 10$. Se desea saber $P(X < 60)$, esto puede aproximarse por métodos numericos como 0.84

````python

from scipy.stats import norm

# Parámetros de la distribución normal
mu = 50  # Media
sigma = 10  # Desviación estándar

# Valores para graficar
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)  # Generar valores de x
pdf = norm.pdf(x, mu, sigma)  # Función de densidad de probabilidad

# Calcular la probabilidad P(X < 60)
probabilidad = norm.cdf(60, mu, sigma)

# Graficar la distribución normal
plt.plot(x, pdf, label='Distribución Normal', color='blue')
plt.fill_between(x, pdf, color='blue', alpha=0.3)
plt.title("Distribución Normal $N(50, 10^2)$")
plt.xlabel("Tiempo de Respuesta (s)")
plt.ylabel("Densidad de Probabilidad")
plt.grid(True)
plt.legend()
plt.show()

# Mostrar la probabilidad
print(f"P(X < 60) = {probabilidad:.4f}")
````

**Nota**

Si $n$ es suficientemente grande $Ber(n, p) \approx N(n p, n p(1-p))$. En la práctica, la aproximación es, por lo general, aceptable cuando $p \in\left(0, \frac{1}{2}\right)$ y $n p(1-p)>9$ o $n p>5$, o si $p \in\left(\frac{1}{2}, 1\right) y$ $n(1-p)>5$.

## Distribución normal estándar Z
Es una variable aleatoria normal con media 0 y varianza 1 . Se suele denotar con la letra $Z$ y su f.d. y f.d.p se suelen denotar como
$$
\varphi(z):=f_Z(z) \quad \Phi(z):=F_Z(z)=p(Z \leq z)
$$

**Teorema**
Sea $X$ una variable aleatoria normal con media $\mu_X$ y varianza $\sigma_X^2$ Entonces, la siguiente variable transformada tiene distribución normal estándar
$$
Z:=\frac{X-\mu_X}{\sigma_X} \sim N\left(0, 1\right)
$$
* Cualquier problema que involucre una variable aleatoria normal puede expresarse entonces como un problema de probabilidad con la distribución normal estándar.
$$
\begin{aligned}
P(X \leq x) & =P\left(X-\mu_X \leq x-\mu_X\right)=P\left(\frac{X-\mu_X}{\sigma_X} \leq \frac{x-\mu_X}{\sigma_X}\right) \\
& =P\left(Z \leq \frac{x-\mu_X}{\sigma_X}\right)=\Phi\left(\frac{x-\mu_X}{\sigma_X}\right)
\end{aligned}
$$

### Ejemplo
El peso de un paquete de galletas en una fábrica sigue una distribución normal con una media de 200 gramos y una desviación estándar de 15 gramos, es decir, $ X \sim N(200, 15^2) $. Se quiere calcular la probabilidad de que un paquete pese entre 190 y 210 gramos, es decir, $ P(190 < X < 210) $.

Primero, estandarizamos:

$$
Z_1 = \frac{190 - 200}{15} = -0.6667, \quad Z_2 = \frac{210 - 200}{15} = 0.6667
$$

La probabilidad es:

$$
P(190 < X < 210) = P(-0.6667 < Z < 0.6667) = P(Z < 0.6667) - P(Z < -0.6667)
$$

Usando la tabla de la normal estándar:

$$
P(Z < 0.6667) \approx 0.745, \quad P(Z < -0.6667) \approx 0.245
$$

Entonces:
$$
P(190 < X < 210) \approx 0.745 - 0.245 = 0.495
$$


## 5. Distribución t de student
La variable $X$ puede tomar cualquier número real. Se denota $X \sim t(n)$, siendo el parámetro $\quad n \in \mathbb{N}^{\dagger}$ : Grados de libertad.

Se dice que una variable aleatoria $X$ tiene distribución t de student si su función de densidad está dada por:
$$f_X(x)=\frac{1}{\sqrt{n \pi}} \Gamma\left(\frac{n+1}{2}\right)\left[\Gamma\left(\frac{n}{2}\right)\right]^{-1}\left(1+\frac{x^2}{n}\right)^{-\left(\frac{n+1}{2}\right)}$$

No existe una expresión analítica para su función de distribución, puede calcularse computacionalmnete

### Propiedades
1. $E[X]=0$.
2. $Var(X)=\frac{n}{n-2}$.

Se muestran algunos posibles resultados de la f. d., dependiendo del valor de los parámetros

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Definir una función para graficar diferentes distribuciones t
def graficar_t_student():
    # Definir diferentes grados de libertad
    grados_libertad = [1, 2, 5, 10, 30]

    # Crear un rango de valores para x
    x = np.linspace(-5, 5, 1000)

    # Graficar la función de densidad de probabilidad para cada valor de grados de libertad
    plt.figure(figsize=(10, 6))
    for df in grados_libertad:
        # Función de densidad de probabilidad (PDF) de la t de Student
        y = t.pdf(x, df)
        plt.plot(x, y, label=f'Grados de libertad = {df}')

    # Configuración del gráfico
    plt.title('Distribución t de Student para diferentes grados de libertad')
    plt.xlabel('x')
    plt.ylabel('Densidad de probabilidad')
    plt.legend()
    plt.grid(True)
    plt.show()

# Llamar a la función
graficar_t_student()
````

## 6. Distribución Chi Cuadrado

La variable $X$ puede tomar valores positivos. Muy útil en inferencia estadística. Sus posibles valores se encuentran en el intervalo $[0, \infty)$. Se denota como $ X \sim \chi^2(n) \stackrel{d}{\approx} \Gamma\left(\frac{n}{2}, \frac{1}{2}\right)$

Se dice que una variable aleatoria $X$ tiene distribución t de student si su función de densidad está dada por:

$$
f_X(x)=\left\{\begin{array}{l}
\frac{(1 / 2)^{n / 2}}{\Gamma(n / 2)} x^{n / 2-1} e^{-x / 2}, x \geq 0 \\
0, \text { en otro caso }
\end{array}\right.
$$

No existe una expresión analítica para su función de distribución, puede calcularse computacionalmnete

### Propiedades
1. $E[X]=n$.
2. $Var(X)=2 n$.

Se muestran algunos posibles resultados de la f. d., dependiendo del valor de los parámetros



````python
from scipy.stats import chi2

# Definir una función para graficar diferentes distribuciones chi-cuadrado
def graficar_chi2():
    # Definir diferentes grados de libertad
    grados_libertad = [2, 5, 10, 20]

    # Crear un rango de valores para x
    x = np.linspace(0, 20, 1000)

    # Graficar la función de densidad de probabilidad para cada valor de grados de libertad
    plt.figure(figsize=(10, 6))
    for df in grados_libertad:
        # Función de densidad de probabilidad (PDF) de la chi-cuadrado
        y = chi2.pdf(x, df)
        plt.plot(x, y, label=f'Grados de libertad = {df}')

    # Configuración del gráfico
    plt.title('Distribución Chi-cuadrado para diferentes grados de libertad')
    plt.xlabel('x')
    plt.ylabel('Densidad de probabilidad')
    plt.legend()
    plt.grid(True)
    plt.show()

# Llamar a la función
graficar_chi2()
````

# 6. Distribución F de Fisher

La **distribución F**, también llamada **distribución F de Fisher-Snedecor**, es una distribución de probabilidad que surge en el análisis de varianza (ANOVA) y en pruebas de hipótesis sobre varianzas.

---

## Definición

La distribución F con **grados de libertad** $d_1$ y $d_2$ se define como la razón de dos variables aleatorias independientes chi-cuadrado normalizadas:

$$
F = \frac{\frac{X_1}{d_1}}{\frac{X_2}{d_2}}
$$

donde:

- $X_1 \sim \chi^2(d_1)$ y $X_2 \sim \chi^2(d_2)$ son variables independientes con distribuciones **chi-cuadrado** con $d_1$ y $d_2$ grados de libertad respectivamente.
- $d_1$ es el **número de grados de libertad del numerador**.
- $d_2$ es el **número de grados de libertad del denominador**.

La variable $F$ se usa para comparar **dos varianzas muestrales** y se aplica en **ANOVA**, **regresión** y **contrastes de hipótesis**.

---

## Función de Densidad de Probabilidad (PDF)

La **función de densidad** de la distribución F es:

$$
f_F(x) = \frac{\Gamma\left(\frac{d_1 + d_2}{2}\right)}{\Gamma\left(\frac{d_1}{2}\right) \Gamma\left(\frac{d_2}{2}\right)}
\cdot \left( \frac{d_1}{d_2} \right)^{\frac{d_1}{2}}
\cdot \frac{x^{\frac{d_1}{2} - 1}}{\left(1 + \frac{d_1}{d_2} x\right)^{\frac{d_1 + d_2}{2}}}
$$

donde $\Gamma(x)$ es la **función gamma**.

La distribución F es **asimétrica positiva** y toma valores en $(0, +\infty)$.

---

## Propiedades Claves

- **Esperanza Matemática**  
  $$
  E[F] = \frac{d_2}{d_2 - 2}, \quad \text{para } d_2 > 2.
  $$

- **Varianza**  
  $$
  \text{Var}(F) = \frac{2 d_2^2 (d_1 + d_2 - 2)}{d_1 (d_2 - 2)^2 (d_2 - 4)}, \quad \text{para } d_2 > 4.
  $$

- **Simetría**  
  La distribución F no es simétrica, sino sesgada hacia la derecha. A medida que $d_1$ y $d_2$ aumentan, la distribución se aproxima a una normal.

- **Relación con la distribución Beta**  
  $$
  F = \frac{d_2}{d_1} \cdot \frac{B\left(\frac{d_1}{2}, \frac{d_2}{2}\right)}{1 - B\left(\frac{d_1}{2}, \frac{d_2}{2}\right)}
  $$

---

## Aplicaciones

La distribución F se usa en:

- **Pruebas de hipótesis sobre varianzas**: Para comparar dos varianzas muestrales.
- **Análisis de varianza (ANOVA)**: Para evaluar diferencias entre grupos.
- **Regresión lineal**: Para contrastar la significancia de los coeficientes del modelo.
- **Inferencia estadística**: En modelos que involucran relaciones entre varianzas.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Grados de libertad
d1, d2 = 5, 10  # Numerador y denominador

# Rango de valores
x = np.linspace(0, 5, 1000)

# Función de densidad de probabilidad (PDF)
pdf = f.pdf(x, d1, d2)

# Graficar
plt.figure(figsize=(8,5))
plt.plot(x, pdf, label=f"F({d1}, {d2})", color="blue")
plt.xlabel("x")
plt.ylabel("Densidad f(x)")
plt.title("Distribución F de Fisher")
plt.legend()
plt.grid()
plt.show()
````

# Relación de las Distribuciones t, F y Chi-Cuadrado con la Distribución Normal

Las distribuciones **t de Student**, **F de Fisher** y **Chi-Cuadrado** están estrechamente relacionadas con la **distribución normal** y desempeñan un papel clave en pruebas de hipótesis y regresión.

---

## 1. Distribución Chi-Cuadrado

### Relación con la Distribución Normal

Si $Z_1, Z_2, ..., Z_k$ son variables aleatorias independientes con distribución normal estándar $N(0,1)$, entonces la **suma de sus cuadrados** sigue una distribución chi-cuadrado con $k$ grados de libertad:

$$
X = Z_1^2 + Z_2^2 + ... + Z_k^2 \sim \chi^2(k)
$$

Esto significa que la distribución **chi-cuadrado surge naturalmente** al sumar cuadrados de variables normales estándar.

---

### Aplicaciones en Regresión y Pruebas de Hipótesis

1. **Prueba de hipótesis sobre varianzas**: Se usa para contrastar si la varianza de una población es igual a un valor específico.
2. **Análisis de regresión**: Se usa en la **prueba de bondad de ajuste** del modelo.
3. **Intervalos de confianza para la varianza**: Se construyen intervalos de confianza para la varianza poblacional con base en la distribución chi-cuadrado.

---

## 2. Distribución t de Student

### Relación con la Distribución Normal

Si $Z$ es una variable normal estándar $N(0,1)$ y $X \sim \chi^2(k)$ con $k$ grados de libertad, entonces la distribución **t de Student** se define como:

$$
T = \frac{Z}{\sqrt{X / k}} \sim t(k)
$$

Donde:
- El numerador $Z$ es una variable normal estándar.
- El denominador es la raíz cuadrada de una chi-cuadrado normalizada.

A medida que $k \to \infty$, la distribución $t(k)$ se aproxima a la normal estándar.

---

### Aplicaciones en Regresión y Pruebas de Hipótesis

1. **Prueba t para la media**: Se usa para contrastar si la media de una población es igual a un valor dado cuando la varianza es desconocida.
2. **Intervalos de confianza para la media**: Se usa cuando el tamaño de muestra es pequeño.
3. **Regresión lineal**: Se usa para evaluar la **significancia de los coeficientes** en modelos de regresión.

---

## 3. Distribución F de Fisher

### Relación con la Distribución Normal

Si $X_1 \sim \chi^2(d_1)$ y $X_2 \sim \chi^2(d_2)$ son variables independientes chi-cuadrado con $d_1$ y $d_2$ grados de libertad, la distribución **F de Fisher** se define como:

$$
F = \frac{\frac{X_1}{d_1}}{\frac{X_2}{d_2}} \sim F(d_1, d_2)
$$

Dado que cada $X_i$ es la suma de cuadrados de variables normales, la distribución **F surge naturalmente** en la comparación de varianzas.

---

### Aplicaciones en Regresión y Pruebas de Hipótesis

1. **Prueba de hipótesis sobre varianzas**: Se usa para comparar dos varianzas muestrales.
2. **Análisis de varianza (ANOVA)**: Se usa para evaluar si existen diferencias entre grupos.
3. **Regresión lineal múltiple**: Se usa en la **prueba global de significancia del modelo**.

---

## 4. Resumen

| Distribución | Relación con la Normal | Aplicaciones |
|-------------|-----------------------|-------------|
| **Chi-Cuadrado** | Suma de normales estándar elevadas al cuadrado | Prueba de varianza, regresión, intervalos de confianza |
| **t de Student** | Normal estándar dividida por la raíz de una chi-cuadrado | Prueba t, regresión, intervalos de confianza |
| **F de Fisher** | Cociente de dos chi-cuadrado normalizadas | ANOVA, regresión múltiple, comparación de varianzas |

---
Estas distribuciones son fundamentales en **pruebas de hipótesis y modelos de regresión** debido a su relación con la distribución normal.
