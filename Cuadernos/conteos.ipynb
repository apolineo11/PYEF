```python
# imports
from itertools import permutations, combinations, product, combinations_with_replacement
import pandas as pd
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
# definición de funciones
def all_integers(*args):
    return all(isinstance(arg, int) for arg in args)

def factorial(n: int) -> int:
    if not isinstance(n, int):
        raise TypeError(f"El argumento {n} debe ser un entero.")
    if n < 0:
        raise ValueError(f"El argumento {n} debe ser mayor o igual a cero.")
    return 1 if n == 0 else n * factorial(n - 1)

def perm(n:int, r:int) -> int:
    if not all_integers(n, r):
        raise TypeError("Los argumentos deben ser enteros.")
    elif n < r:
        raise ValueError(f"El argumento {n} debe ser mayor o igual que el argumento {r}.")
    else:
        return factorial(n)//factorial(n-r)

def comb(n:int, r:int) -> int:
    return perm(n,r)//factorial(r)
```

# Principio de multiplicación y adición

Un restaurantes ofrece 2 menús, uno para personas que NO comen carne y otro para personas que SÍ comen carne.
Un comensal solo puede armar su plato escogiendo los componentes de uno solo de los menús.

- ¿De cuantas formas puede una persona que SÍ come carne armar su plato?
- ¿De cuantas formas puede una persona que NO come carne armar su plato?
- ¿Cuantas opciones tiene cualquier persona para armar su plato?

```python
# Principio de multiplicación y adición
# menú con carne
bebidas = ["Limonada", "Coca-Cola", "Cerveza", "Jugo Tamarindo", "Jugo Guanabana"]
carnes = ["Lomo", "Costilla", "Pechuga", "Mojarra"]
acompanantes = ["Papa", "Yuca", "Patacón"]
salsas = ["BBQ", "Guacamole", "Ají", "Mostaza", "Rosada", "Chimichurri"]

# menú sin carne
bebidas_nc = ["Jugo Tamarindo", "Jugo Guanabana", "Té Frío", "Limonada"]
proteinas_nc = ["Tofu", "Setas", "Hummus", "Lentejas"]
cereales_nc = ["Arroz Integral", "Quinoa"]
ensaladas_nc = ["Aguacate", "Roja", "Espinaca", "Berenjena"]
```

```python
# De cuantas formas puede armar su plato una persona que come carne

carne_opciones = list(product(bebidas, carnes, acompanantes, salsas))
print(f"""
La persona que coma carne tiene {len(carne_opciones)} opciones para escoger
"""
)
for carne_opcion in carne_opciones:
    print(carne_opcion)
```

```python
# De cuantas formas puede armar su plato una persona que NO come carne
no_carne_opciones = list(product(bebidas_nc, proteinas_nc, cereales_nc, ensaladas_nc))
print(f"""
La persona que NO coma carne tiene {len(no_carne_opciones)} opciones para escoger
"""
)
for no_carne_opcion in no_carne_opciones:
    print(no_carne_opcion)
```

```python
# Principio de adición
total_menu = carne_opciones + no_carne_opciones
print(f"""
En general una persona que vaya al restaurante puede elegir uno de los platos con carne o uno los platos sin carne.
Por ende, puede elegir entre {len(total_menu)} platos.
""")
```

# Permutaciones

Supongamos que un comensal que SÍ come carne ya ha ordenado su plato, tal persona lleva un estricto orden al comer, es decir si elige empezar por un elemento de su plato lo termina, luego empieza por otro elemento hasta que haya finalizado.

- ¿En cuantas maneras se puede comer el plato si solo come un componente a la vez?
- ¿En cuantas manera si deja la bedida para el último lugar

```python
# Persona que come carne
rng = np.random.default_rng(73)
opcion_carne = rng.integers(low=0, high=len(carne_opciones)-1, size=1)[0]
carne_0 = carne_opciones[opcion_carne]
print(f"""
El comensal ha elegido {carne_0}
""")
orden_carne_0 = list(permutations(carne_0))
print(f"""
El comensal puede comerse su plato en {len(orden_carne_0)} ordenes.
""")
for orden in orden_carne_0:
    print(orden)
```

```python
ultimo_bebida = list(product(list(permutations(carne_0[1:])), [carne_0[0]]))
print(f"""
Si el comensal siempre deja la bebida en último lugar tiene {len(ultimo_bebida)} formas de comer su almuerzo
""")
for item in ultimo_bebida:
    print(item)
```

Un estudiante ha recibido recomendaciones para ver 5 series el fin de semana, pero el solo puede dedicar 3 horas a esta actividad porque debe estudiar para el parcial de PYEF. Si en promedio cada capítulo dura una hora (sin importar la serie), el estudiante decide ver tres capítulos.

- ¿De cuantas formas puede agendar su fin de semana para ver 3 capítulos de series diferentes?

```python
# Series

series = ["The Studio", "Separación (Severance)", "Adolescencia", "Daredevil: Born Again", "The Pitt"]
fin_semana_series_orden = list(permutations(series, 3))
print(f"""
El estudiante tiene {len(fin_semana_series_orden)} formas de agendar su fin de semana para ver 3 capítulos
""")
for agenda in fin_semana_series_orden:
    print(agenda)
```

# Combinaciones

Al estudiante le tiene sin cuidado el orden en el que va ver los tres capitulos de diferentes series

- ¿De cuantas formas puede ver los tres capitulos?

```python
fin_semana_series_no_orden = list(combinations(series, 3))
print(f"""
El estudiante tiene {len(fin_semana_series_no_orden)} formas de agendar su fin de semana para ver 3 capítulos
""")
for agenda in fin_semana_series_no_orden:
    print(agenda)
```

# Combinaciones con repetición

El estudiante ha decidido que no necesariamente va ver un capítulo de cada serie que elija. Sino que puede ver más de un capítulo de una serie, de tal forma que puede llegar a ver 3 capítulos de una sola serie.

```python
fin_semana_series_repeticion = list(combinations_with_replacement(series, 3))
print(f"""
El estudiante tiene {len(fin_semana_series_repeticion)} formas de agendar su fin de semana para ver 3 capítulos
""")
for agenda in fin_semana_series_repeticion:
    print(agenda)
```

# Comportamientos

```python
# Gráfico permutaciones con n fijo
n = 6
plt.figure(figsize=(8, 5))
for ri in range(0,n+1):
    print(f"{n}P{ri}={perm(n,ri)}")
    plt.plot([ri, ri], [0, perm(n,ri)], color="skyblue", linewidth=2)
    plt.plot(ri, perm(n,ri), "o", color="blue")

plt.xlabel('r')
plt.ylabel('nPr')
plt.title(f'Permutación de {n} elementos')
plt.grid(True)
plt.show()
```

```python
# Gráfico Combinaciones con n fijo
n = 6
plt.figure(figsize=(8, 5))

for ri in range(0,n+1):
    print(f"{n}P{ri}={comb(n,ri)}")
    plt.plot([ri, ri], [0, comb(n,ri)], color="skyblue", linewidth=2)
    plt.plot(ri, comb(n,ri), "o", color="blue")

plt.xlabel('r')
plt.ylabel('nPr')
plt.title(f'Combinación de {n} elementos')
plt.grid(True)
plt.show()
```

```python
# Gráfico de permutaciones y combinaciones con r fijo
n = 6
r = 3
n_values = list(range(0, n+1))
perms = [math.perm(n, r) for n in n_values]
combs = [math.comb(n, r) for n in n_values]

plt.figure(figsize=(8, 5))
plt.plot(n_values, perms, label='Permutaciones')
plt.plot(n_values, combs, label='Combinaciones')
plt.xlabel('n')
plt.ylabel('Cantidad')
plt.title('Comparación entre permutaciones y combinaciones')
plt.legend()
plt.grid(True)
plt.show()
```
